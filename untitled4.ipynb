```python
# ============================================================
# improved_gait_lstm_variable_length_5fold_cv.py
# 可変長 LSTM + Δt 分割 + 5Fold クロスバリデーション + NRMSE + グラフ
# ============================================================

import os
import random
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

from sklearn.model_selection import KFold

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
from torch.nn.utils.rnn import (
    pad_sequence, pack_padded_sequence, pad_packed_sequence
)


# ============================================================
# 0. reproducibility
# ============================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


# ============================================================
# 1. 設定
# ============================================================
file_path = "/content/drive/MyDrive/t-teramae/new_wf_tr_extraction_selected.csv"
encoding = "cp932"
low_memory = False

batch_size = 8
num_epochs = 40
initial_lr = 5e-4

hidden_dim = 96
num_layers = 2
dropout = 0.2
grad_clip = 1.0

input_keys = [ "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm", "左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm", "左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm", "左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm", "左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm", "左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm", "右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm", "右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm", "右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm", "右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm", "右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm", "右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm", "左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm", "左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm", "左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm", "左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm", "左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm", "左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm", "右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm", "右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm", "右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm", "右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm", "PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm", "股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm", "肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm", "体重心Xcm","体重心Ycm","体重心Zcm" ]
output_keys = [ "左足Fzkg","左足Fxkg","左足Fykg", "右足Fzkg","右足Fxkg","右足Fykg" ]

param_keys = ["年齢", "身長", "体重", "ID", "p_side"]



# ============================================================
# 2. CSV 読み込み
# ============================================================
df = pd.read_csv(file_path, encoding=encoding, low_memory=low_memory)
print("CSV 読み込み完了:", len(df))

# 時間列を自動検出
possible_time_cols = ["time", "Time", "timestamp", "時間", "TIME"]
TIME_COL = next((c for c in possible_time_cols if c in df.columns), None)
if TIME_COL is None:
    raise KeyError("時間列が見つかりません")

df[TIME_COL] = df[TIME_COL].astype(float)


# ============================================================
# 3. Δt によるトライアル分割
# ============================================================
tol = 5e-4
dt = df[TIME_COL].diff().fillna(0.0).values
is_regular = np.isclose(dt, 0.001, atol=tol)

trial_id = []
cur = 0
for i, reg in enumerate(is_regular):
    if i == 0:
        trial_id.append(cur)
        continue
    if not reg:
        cur += 1
    trial_id.append(cur)

df["trial_id"] = np.array(trial_id)
num_trials = df["trial_id"].max() + 1
print("検出されたトライアル:", num_trials)



# ============================================================
# 4. トライアルごとにシーケンス抽出
# ============================================================
raw_X, raw_y = [], []
MIN_LEN = 5
bad_trials = []

for tid in range(num_trials):
    chunk = df[df["trial_id"] == tid]
    if len(chunk) < MIN_LEN:
        continue

    if chunk[input_keys].isna().any().any() or chunk[output_keys].isna().any().any():
        continue

    raw_X.append(chunk[input_keys].values.astype(np.float32))
    raw_y.append(chunk[output_keys].values.astype(np.float32))

print("有効トライアル:", len(raw_X))



# ============================================================
# 5. Dataset / DataLoader
# ============================================================
class VarSeqDataset(Dataset):
    def __init__(self, X_list, y_list):
        self.X = X_list
        self.y = y_list
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        x = torch.tensor(self.X[idx])
        y = torch.tensor(self.y[idx])
        return x, y, x.size(0)

def collate_fn(batch):
    xs, ys, lengths = zip(*batch)
    xs_pad = pad_sequence(xs, batch_first=True)
    ys_pad = pad_sequence(ys, batch_first=True)
    lengths = torch.tensor(lengths)
    return xs_pad, ys_pad, lengths



# ============================================================
# 6. モデル
# ============================================================
class LSTMVarModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers,
                            batch_first=True, dropout=dropout)
        self.ln = nn.LayerNorm(hidden_dim)
        self.fc1 = nn.Linear(hidden_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()
        self.drop = nn.Dropout(dropout)

    def forward(self, x_pad, lengths):
        packed = pack_padded_sequence(x_pad, lengths.cpu(),
                                      batch_first=True, enforce_sorted=False)
        packed_out, _ = self.lstm(packed)
        out, _ = pad_packed_sequence(packed_out, batch_first=True)

        # layernorm only for valid rows
        B, Lmax, H = out.shape
        ar = torch.arange(Lmax, device=out.device)[None, :]
        mask = (ar < lengths[:, None])
        flat = out.reshape(-1, H)

        valid_idx = mask.reshape(-1).nonzero(as_tuple=False).squeeze(1)

        flat_norm = flat.clone()

        if valid_idx.numel() > 0:
            # ★ AMP 互換：dtype を flat_norm に揃える
            ln_out = self.ln(flat[valid_idx]).to(flat_norm.dtype)
            flat_norm[valid_idx] = ln_out

        out = flat_norm.reshape(B, Lmax, H)

        out = self.relu(self.fc1(out))
        out = self.drop(out)
        out = self.fc2(out)
        return out


input_dim = raw_X[0].shape[1]
output_dim = raw_y[0].shape[1]



# ============================================================
# 7. 5Fold クロスバリデーション
# ============================================================
kf = KFold(n_splits=5, shuffle=True, random_state=SEED)

fold_results = []

for fold, (train_idx, val_idx) in enumerate(kf.split(raw_X)):
    print()
    print("="*60)
    print(f" Fold {fold+1} / 5")
    print("="*60)

    X_train = [raw_X[i] for i in train_idx]
    y_train = [raw_y[i] for i in train_idx]
    X_val = [raw_X[i] for i in val_idx]
    y_val = [raw_y[i] for i in val_idx]

    # --- 標準化（Fold 内） ---
    mean = np.concatenate(X_train).mean(axis=0)
    std = np.concatenate(X_train).std(axis=0)
    std[std == 0] = 1.0

    X_train = [(x - mean)/std for x in X_train]
    X_val = [(x - mean)/std for x in X_val]

    train_loader = DataLoader(VarSeqDataset(X_train, y_train),
                              batch_size=batch_size, shuffle=True,
                              collate_fn=collate_fn)
    val_loader = DataLoader(VarSeqDataset(X_val, y_val),
                            batch_size=batch_size, shuffle=False,
                            collate_fn=collate_fn)

    # --- モデル ---
    model = LSTMVarModel(input_dim, hidden_dim, output_dim,
                         num_layers=num_layers, dropout=dropout).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=initial_lr)
    scaler = GradScaler()

    # ----------------------
    # 学習
    # ----------------------
    for ep in range(1, num_epochs+1):
        model.train()
        for Xb, yb, Lb in train_loader:
            Xb, yb, Lb = Xb.to(device), yb.to(device), Lb.to(device)
            opt.zero_grad()

            with autocast():
                pred = model(Xb, Lb)
                # Mask MSE
                B, Lmax, C = pred.shape
                ar = torch.arange(Lmax, device=device)[None, :]
                mask = (ar < Lb[:, None]).unsqueeze(-1)
                se = ((pred - yb)**2) * mask
                denom = mask.sum()*C
                loss = se.sum() / denom

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            scaler.step(opt)
            scaler.update()

        if ep % 10 == 0:
            print(f"  Epoch {ep} done.")


    # ----------------------
    # 検証
    # ----------------------
    model.eval()
    preds = []
    tgts = []

    with torch.no_grad():
        for Xv, yv, Lv in val_loader:
            Xv, yv, Lv = Xv.to(device), yv.to(device), Lv.to(device)
            pred = model(Xv, Lv)

            pred = pred.cpu().numpy()
            yv = yv.cpu().numpy()
            Lv = Lv.cpu().numpy()

            for i in range(pred.shape[0]):
                L = Lv[i]
                preds.append(pred[i, :L])
                tgts.append(yv[i, :L])


    pred_concat = np.concatenate(preds)
    tgt_concat = np.concatenate(tgts)

    range_c = tgt_concat.max(axis=0) - tgt_concat.min(axis=0)
    range_c[range_c == 0] = 1.0

    ch_nrmse = np.sqrt(((pred_concat - tgt_concat)**2).mean(axis=0)) / range_c
    fold_nrmse = np.sqrt(((pred_concat - tgt_concat)**2).mean()) / (tgt_concat.max()-tgt_concat.min())

    fold_results.append(fold_nrmse)
    print(f"  Fold {fold+1} NRMSE = {fold_nrmse:.6f}")

    # ------------------------------------------------------
    # 検証トライアルのグラフ表示（お望み通りすべて表示）
    # ------------------------------------------------------
    for i, (p, t) in enumerate(zip(preds, tgts)):
        plt.figure(figsize=(12,5))
        for ch in range(p.shape[1]):
            plt.plot(t[:, ch], "--", label=f"tgt {output_keys[ch]}")
            plt.plot(p[:, ch], label=f"pred {output_keys[ch]}")
        plt.title(f"Fold {fold+1} : Validation Trial {i}")
        plt.legend()
        plt.tight_layout()
        plt.savefig(f"/content/drive/MyDrive/t-teramae/fig/fold_{fold+1}_trial_{i}.png")
        plt.close()


print("="*60)
print(" 5Fold CV 結果")
print("="*60)
for i, v in enumerate(fold_results):
    print(f" Fold {i+1} NRMSE = {v:.6f}")

print(" 平均 NRMSE =", np.mean(fold_results))
```

```python
# ============================================================
# improved_gait_lstm_variable_length_5fold_cv_fixed.py
# 修正版: 可変長 LSTM + Δt 分割 + 5Fold クロスバリデーション + NRMSE + グラフ
# - fixes: trial id mapping, per-channel NRMSE, filenames, small robustness tweaks
# ============================================================

import os
import random
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

from sklearn.model_selection import KFold

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
from torch.nn.utils.rnn import (
    pad_sequence, pack_padded_sequence, pad_packed_sequence
)

# ============================================================
# 0. reproducibility
# ============================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ============================================================
# 1. 設定
# ============================================================

file_path = "/content/drive/MyDrive/t-teramae/new_wf_tr_extraction_selected.csv"
encoding = "cp932"
low_memory = False

batch_size = 8
num_epochs = 40
initial_lr = 5e-4

hidden_dim = 96
num_layers = 2
dropout = 0.2
grad_clip = 1.0

input_keys = [ "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm", "左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm", "左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm", "左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm", "左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm", "左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm", "右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm", "右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm", "右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm", "右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm", "右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm", "右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm", "左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm", "左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm", "左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm", "左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm", "左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm", "左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm", "右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm", "右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm", "右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm", "右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm", "PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm", "股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm", "肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm", "体重心Xcm","体重心Ycm","体重心Zcm" ]
output_keys = [ "左足Fzkg","左足Fxkg","左足Fykg", "右足Fzkg","右足Fxkg","右足Fykg" ]
param_keys = ["年齢", "身長", "体重", "ID", "p_side"]

# 保存ディレクトリ作成
save_dir = "/content/drive/MyDrive/t-teramae/results"
os.makedirs(save_dir, exist_ok=True)
fig_dir = "/content/drive/MyDrive/t-teramae/fig"
os.makedirs(fig_dir, exist_ok=True)

# ============================================================
# 2. CSV 読み込み
# ============================================================
df = pd.read_csv(file_path, encoding=encoding, low_memory=low_memory)
print("CSV 読み込み完了:", len(df))

# 時間列を自動検出
possible_time_cols = ["time", "Time", "timestamp", "時間", "TIME"]
TIME_COL = next((c for c in possible_time_cols if c in df.columns), None)
if TIME_COL is None:
    raise KeyError("時間列が見つかりません")

df[TIME_COL] = df[TIME_COL].astype(float)

# ============================================================
# 3. Δt によるトライアル分割
# ============================================================
tol = 5e-4   # 必要ならここを少し大きくする（例: 1e-3）ことを検討
dt = df[TIME_COL].diff().fillna(0.0).values
is_regular = np.isclose(dt, 0.001, atol=tol)

trial_id = []
cur = 0
for i, reg in enumerate(is_regular):
    if i == 0:
        trial_id.append(cur)
        continue
    if not reg:
        cur += 1
    trial_id.append(cur)

df["trial_id"] = np.array(trial_id)
num_trials = int(df["trial_id"].max() + 1)
print("検出されたトライアル:", num_trials)

# ============================================================
# 4. トライアルごとにシーケンス抽出
#    → ここで元の trial id を保持する（raw_trial_ids）
# ============================================================
raw_X, raw_y, raw_trial_ids = [], [], []
MIN_LEN = 5

for tid in range(num_trials):
    chunk = df[df["trial_id"] == tid]
    if len(chunk) < MIN_LEN:
        continue
    if chunk[input_keys].isna().any().any() or chunk[output_keys].isna().any().any():
        continue

    raw_X.append(chunk[input_keys].values.astype(np.float32))
    raw_y.append(chunk[output_keys].values.astype(np.float32))
    raw_trial_ids.append(int(tid))

print("有効トライアル:", len(raw_X))

if len(raw_X) == 0:
    raise RuntimeError("有効なトライアルが0件です。入力データ / フィルタ条件を確認してください。")

# ============================================================
# Dataset / DataLoader
# ============================================================
class VarSeqDataset(Dataset):
    def __init__(self, X_list, y_list):
        self.X = X_list
        self.y = y_list
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        x = torch.tensor(self.X[idx])
        y = torch.tensor(self.y[idx])
        return x, y, x.size(0)

def collate_fn(batch):
    xs, ys, lengths = zip(*batch)
    xs_pad = pad_sequence(xs, batch_first=True)
    ys_pad = pad_sequence(ys, batch_first=True)
    lengths = torch.tensor(lengths)
    return xs_pad, ys_pad, lengths

# ============================================================
# 6. モデル
# ============================================================
class LSTMVarModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers,
                            batch_first=True, dropout=dropout)
        self.ln = nn.LayerNorm(hidden_dim)
        self.fc1 = nn.Linear(hidden_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()
        self.drop = nn.Dropout(dropout)

    def forward(self, x_pad, lengths):
        packed = pack_padded_sequence(x_pad, lengths.cpu(),
                                      batch_first=True, enforce_sorted=False)
        packed_out, _ = self.lstm(packed)
        out, _ = pad_packed_sequence(packed_out, batch_first=True)

        # out: (B, Lmax, H)
        B, Lmax, H = out.shape
        ar = torch.arange(Lmax, device=out.device)[None, :]
        mask = (ar < lengths[:, None])  # (B, Lmax)
        flat = out.reshape(-1, H)
        valid_idx = mask.reshape(-1).nonzero(as_tuple=False).squeeze(1)

        flat_norm = flat.clone()
        if valid_idx.numel() > 0:
            ln_out = self.ln(flat[valid_idx]).to(flat_norm.dtype)
            flat_norm[valid_idx] = ln_out

        out = flat_norm.reshape(B, Lmax, H)
        out = self.relu(self.fc1(out))
        out = self.drop(out)
        out = self.fc2(out)  # (B, Lmax, output_dim)
        return out

input_dim = raw_X[0].shape[1]
output_dim = raw_y[0].shape[1]

# ============================================================
# 7. 5Fold クロスバリデーション
# ============================================================
kf = KFold(n_splits=5, shuffle=True, random_state=SEED)
fold_results = []
per_fold_channel_nrmse = []

# We'll need mapping from "index in raw_X" -> original trial id -> subject ID
trial_to_subject = df.groupby("trial_id")["ID"].first().to_dict()

for fold, (train_idx, val_idx) in enumerate(kf.split(raw_X)):
    print()
    print("="*60)
    print(f" Fold {fold+1} / 5")
    print("="*60)

    # ---- データ分割 ----
    X_train = [raw_X[i] for i in train_idx]
    y_train = [raw_y[i] for i in train_idx]
    X_val = [raw_X[i] for i in val_idx]
    y_val = [raw_y[i] for i in val_idx]

    # ---- 標準化 ----
    # Concatenate along time dimension to compute per-feature mean/std
    concat_Xtrain = np.concatenate(X_train, axis=0)
    mean = concat_Xtrain.mean(axis=0)
    std = concat_Xtrain.std(axis=0)
    std[std == 0] = 1.0

    X_train = [(x - mean)/std for x in X_train]
    X_val = [(x - mean)/std for x in X_val]

    train_loader = DataLoader(VarSeqDataset(X_train, y_train),
                              batch_size=batch_size, shuffle=True,
                              collate_fn=collate_fn)
    val_loader = DataLoader(VarSeqDataset(X_val, y_val),
                            batch_size=batch_size, shuffle=False,
                            collate_fn=collate_fn)

    # ---- モデル ----
    model = LSTMVarModel(input_dim, hidden_dim, output_dim,
                         num_layers=num_layers, dropout=dropout).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=initial_lr)
    scaler = GradScaler()

    # ============================================================
    # 学習
    # ============================================================
    for ep in range(1, num_epochs+1):
        model.train()
        for Xb, yb, Lb in train_loader:
            Xb, yb, Lb = Xb.to(device), yb.to(device), Lb.to(device)
            opt.zero_grad()

            with autocast():
                pred = model(Xb, Lb)  # (B, Lmax, C)
                B, Lmax, C = pred.shape
                ar = torch.arange(Lmax, device=device)[None, :]
                mask = (ar < Lb[:, None]).unsqueeze(-1)  # (B, Lmax, 1)
                se = ((pred - yb)**2) * mask  # masked squared error
                denom = mask.sum() * C  # total valid elements
                loss = se.sum() / denom

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            scaler.step(opt)
            scaler.update()

        if ep % 10 == 0:
            print(f"  Epoch {ep} done.")

    # ============================================================
    # 検証
    # ============================================================
    model.eval()
    preds = []
    tgts = []
    val_raw_trial_indices = []  # store mapping to raw trial index (index in raw_X)
    with torch.no_grad():
        for batch_i, (Xv, yv, Lv) in enumerate(val_loader):
            Xv, yv, Lv = Xv.to(device), yv.to(device), Lv.to(device)
            pred = model(Xv, Lv)

            pred = pred.cpu().numpy()
            yv = yv.cpu().numpy()
            Lv = Lv.cpu().numpy()

            for i in range(pred.shape[0]):
                L = int(Lv[i])
                preds.append(pred[i, :L])
                tgts.append(yv[i, :L])
                # find the corresponding raw index: val_idx lists raw indices in order
                # We will map these sequentially: keep track by count
                val_raw_trial_indices.append(int(val_idx[len(val_raw_trial_indices)]))

    # Concatenate for metric computation
    pred_concat = np.concatenate(preds, axis=0)  # (N_total, C)
    tgt_concat = np.concatenate(tgts, axis=0)    # (N_total, C)

    # per-channel NRMSE: RMSE(channel) / range(channel)
    channel_rmse = np.sqrt(((pred_concat - tgt_concat)**2).mean(axis=0))  # shape (C,)
    channel_range = tgt_concat.max(axis=0) - tgt_concat.min(axis=0)
    channel_range[channel_range == 0] = 1.0
    channel_nrmse = channel_rmse / channel_range  # shape (C,)

    fold_nrmse = float(channel_nrmse.mean())
    fold_results.append(fold_nrmse)
    per_fold_channel_nrmse.append(channel_nrmse)

    print(f"  Fold {fold+1} channel NRMSE = {channel_nrmse}")
    print(f"  Fold {fold+1} NRMSE (mean over channels) = {fold_nrmse:.6f}")

    # ============================================================
    # NRMSE 計算に使用した元データを保存（ID 情報付き）
    # ============================================================

    # raw_trial_ids maps index in raw_X -> original trial_id
    # val_idx contains indices into raw_X; so val_raw_trial_indices is list of original raw indices used in order
    rows = []
    # val_raw_trial_indices length equals number of validation samples
    for sample_i, (p_arr, t_arr) in enumerate(zip(preds, tgts)):
        L = len(p_arr)
        raw_index = int(val_idx[sample_i])  # index into raw_X
        orig_trial_id = raw_trial_ids[raw_index]  # original trial id in df
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        for t_idx in range(L):
            row = {"raw_index": raw_index, "trial_id": orig_trial_id, "ID": subject_id, "time_idx": t_idx}
            for ch in range(output_dim):
                row[f"pred_{output_keys[ch]}"] = float(p_arr[t_idx, ch])
                row[f"tgt_{output_keys[ch]}"] = float(t_arr[t_idx, ch])
            rows.append(row)

    df_nrmse = pd.DataFrame(rows)
    df_nrmse.to_csv(f"{save_dir}/fold_{fold+1}_nrmse_data.csv", index=False)

    # ============================================================
    # 各検証トライアルの CSV 保存（ファイル名に raw trial id と subject ID を含める）
    # ============================================================
    # preds/tgts order corresponds to order of val_idx
    for i, (p, t) in enumerate(zip(preds, tgts)):
        raw_index = int(val_idx[i])
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        df_out = pd.DataFrame()
        df_out["time_index"] = np.arange(len(p))
        for ch in range(output_dim):
            df_out[f"pred_{output_keys[ch]}"] = p[:, ch]
            df_out[f"tgt_{output_keys[ch]}"] = t[:, ch]

        fname = f"{save_dir}/fold_{fold+1}_rawidx_{raw_index}_trialid_{orig_trial_id}_ID_{subject_id}.csv"
        df_out.to_csv(fname, index=False)

    # ============================================================
    # グラフ保存（ファイル名に raw trial id を含める）
    # ============================================================
    for i, (p, t) in enumerate(zip(preds, tgts)):
        raw_index = int(val_idx[i])
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        plt.figure(figsize=(12,5))
        for ch in range(p.shape[1]):
            plt.plot(t[:, ch], "--", label=f"tgt {output_keys[ch]}")
            plt.plot(p[:, ch], label=f"pred {output_keys[ch]}")
        plt.title(f"Fold {fold+1} : Validation Trial {i} (rawidx {raw_index}, trial {orig_trial_id}, ID {subject_id})")
        plt.legend()
        plt.tight_layout()
        plt.savefig(f"{fig_dir}/fold_{fold+1}_rawidx_{raw_index}_trialid_{orig_trial_id}_ID_{subject_id}.png")
        plt.close()

# ============================================================
# クロスバリデーション結果の保存
# ============================================================
cv_df = pd.DataFrame({
    "fold": list(range(1, len(fold_results)+1)),
    "NRMSE_mean_over_channels": fold_results
})
cv_df.loc[len(cv_df)] = ["mean", np.mean(fold_results)]
cv_df.to_csv(f"{save_dir}/cv_results.csv", index=False)

# Optionally save per-fold per-channel NRMSE table
channel_cols = [f"NRMSE_{k}" for k in output_keys]
channel_table = pd.DataFrame(np.vstack(per_fold_channel_nrmse), columns=channel_cols)
channel_table.insert(0, "fold", list(range(1, len(per_fold_channel_nrmse)+1)))
channel_table.loc[len(channel_table)] = ["mean"] + list(channel_table[channel_cols].mean(axis=0))
channel_table.to_csv(f"{save_dir}/cv_channel_nrmse.csv", index=False)

print("="*60)
print(" 5Fold CV 結果")
print("="*60)
for i, v in enumerate(fold_results):
    print(f" Fold {i+1} NRMSE = {v:.6f}")
print(" 平均 NRMSE =", np.mean(fold_results))
print(f"\n→ 保存先: {save_dir}/cv_results.csv")
```

```python
# ============================================================
# improved_gait_lstm_variable_length_5fold_cv_fixed.py
# 修正版: 可変長 LSTM + Δt 分割 + 5Fold クロスバリデーション + NRMSE + グラフ
# - fixes: trial id mapping, per-channel NRMSE, filenames, small robustness tweaks
# ============================================================

import os
import random
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

from sklearn.model_selection import KFold

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
from torch.nn.utils.rnn import (
    pad_sequence, pack_padded_sequence, pad_packed_sequence
)

# ============================================================
# 0. reproducibility
# ============================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ============================================================
# 1. 設定
# ============================================================

file_path = "/content/new_wf_tr_extraction_selected.csv"
encoding = "cp932"
low_memory = False

batch_size = 8
num_epochs = 40
initial_lr = 5e-4

hidden_dim = 96
num_layers = 2
dropout = 0.2
grad_clip = 1.0

input_keys = [ "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm", "左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm", "左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm", "左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm", "左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm", "左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm", "右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm", "右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm", "右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm", "右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm", "右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm", "右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm", "左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm", "左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm", "左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm", "左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm", "左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm", "左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm", "右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm", "右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm", "右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm", "右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm", "PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm", "股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm", "肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm", "体重心Xcm","体重心Ycm","体重心Zcm" ]
output_keys = [ "左足Fzkg","左足Fxkg","左足Fykg", "右足Fzkg","右足Fxkg","右足Fykg" ]
param_keys = ["年齢", "身長", "体重", "ID", "p_side"]

# 保存ディレクトリ作成
save_dir = "/content/drive/MyDrive/t-teramae/results"
os.makedirs(save_dir, exist_ok=True)
fig_dir = "/content/drive/MyDrive/t-teramae/fig"
os.makedirs(fig_dir, exist_ok=True)

# ============================================================
# 2. CSV 読み込み
# ============================================================
df = pd.read_csv(file_path, encoding=encoding, low_memory=low_memory)
print("CSV 読み込み完了:", len(df))

# 時間列を自動検出
possible_time_cols = ["time", "Time", "timestamp", "時間", "TIME"]
TIME_COL = next((c for c in possible_time_cols if c in df.columns), None)
if TIME_COL is None:
    raise KeyError("時間列が見つかりません")

df[TIME_COL] = df[TIME_COL].astype(float)

# ============================================================
# 3. Δt によるトライアル分割
# ============================================================
tol = 5e-4   # 必要ならここを少し大きくする（例: 1e-3）ことを検討
dt = df[TIME_COL].diff().fillna(0.0).values
is_regular = np.isclose(dt, 0.001, atol=tol)

trial_id = []
cur = 0
for i, reg in enumerate(is_regular):
    if i == 0:
        trial_id.append(cur)
        continue
    if not reg:
        cur += 1
    trial_id.append(cur)

df["trial_id"] = np.array(trial_id)
num_trials = int(df["trial_id"].max() + 1)
print("検出されたトライアル:", num_trials)

# ============================================================
# 4. トライアルごとにシーケンス抽出
#    → ここで元の trial id を保持する（raw_trial_ids）
# ============================================================
raw_X, raw_y, raw_trial_ids = [], [], []
MIN_LEN = 5

for tid in range(num_trials):
    chunk = df[df["trial_id"] == tid]
    if len(chunk) < MIN_LEN:
        continue
    if chunk[input_keys].isna().any().any() or chunk[output_keys].isna().any().any():
        continue

    raw_X.append(chunk[input_keys].values.astype(np.float32))
    raw_y.append(chunk[output_keys].values.astype(np.float32))
    raw_trial_ids.append(int(tid))

print("有効トライアル:", len(raw_X))

if len(raw_X) == 0:
    raise RuntimeError("有効なトライアルが0件です。入力データ / フィルタ条件を確認してください。")

# ============================================================
# Dataset / DataLoader
# ============================================================
class VarSeqDataset(Dataset):
    def __init__(self, X_list, y_list):
        self.X = X_list
        self.y = y_list
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        x = torch.tensor(self.X[idx])
        y = torch.tensor(self.y[idx])
        return x, y, x.size(0)

def collate_fn(batch):
    xs, ys, lengths = zip(*batch)
    xs_pad = pad_sequence(xs, batch_first=True)
    ys_pad = pad_sequence(ys, batch_first=True)
    lengths = torch.tensor(lengths)
    return xs_pad, ys_pad, lengths

# ============================================================
# 6. モデル
# ============================================================
class LSTMVarModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=2, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=num_layers,
                            batch_first=True, dropout=dropout)
        self.ln = nn.LayerNorm(hidden_dim)
        self.fc1 = nn.Linear(hidden_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()
        self.drop = nn.Dropout(dropout)

    def forward(self, x_pad, lengths):
        packed = pack_padded_sequence(x_pad, lengths.cpu(),
                                      batch_first=True, enforce_sorted=False)
        packed_out, _ = self.lstm(packed)
        out, _ = pad_packed_sequence(packed_out, batch_first=True)

        # out: (B, Lmax, H)
        B, Lmax, H = out.shape
        ar = torch.arange(Lmax, device=out.device)[None, :]
        mask = (ar < lengths[:, None])  # (B, Lmax)
        flat = out.reshape(-1, H)
        valid_idx = mask.reshape(-1).nonzero(as_tuple=False).squeeze(1)

        flat_norm = flat.clone()
        if valid_idx.numel() > 0:
            ln_out = self.ln(flat[valid_idx]).to(flat_norm.dtype)
            flat_norm[valid_idx] = ln_out

        out = flat_norm.reshape(B, Lmax, H)
        out = self.relu(self.fc1(out))
        out = self.drop(out)
        out = self.fc2(out)  # (B, Lmax, output_dim)
        return out

input_dim = raw_X[0].shape[1]
output_dim = raw_y[0].shape[1]

# ============================================================
# 7. GroupKFold クロスバリデーション（被験者 ID ごとにグループ分割）
# ============================================================
from sklearn.model_selection import GroupKFold

fold_results = []
per_fold_channel_nrmse = []

# We'll need mapping from "index in raw_X" -> original trial id -> subject ID
trial_to_subject = df.groupby("trial_id")["ID"].first().to_dict()

# raw_X の各要素（トライアル）の subject ID をグループラベルにする
groups = np.array([trial_to_subject[raw_trial_ids[i]] for i in range(len(raw_X))])

gkf = GroupKFold(n_splits=5)

for fold, (train_idx, val_idx) in enumerate(gkf.split(raw_X, groups=groups)):
    print()
    print("="*60)
    print(f" Fold {fold+1} / 5  (GroupKFold)")
    print("="*60)

    # ---- データ分割 ----
    X_train = [raw_X[i] for i in train_idx]
    y_train = [raw_y[i] for i in train_idx]
    X_val   = [raw_X[i] for i in val_idx]
    y_val   = [raw_y[i] for i in val_idx]

    # ---- 標準化 ----
    concat_Xtrain = np.concatenate(X_train, axis=0)
    mean = concat_Xtrain.mean(axis=0)
    std = concat_Xtrain.std(axis=0)
    std[std == 0] = 1.0

    X_train = [(x - mean)/std for x in X_train]
    X_val = [(x - mean)/std for x in X_val]

    train_loader = DataLoader(VarSeqDataset(X_train, y_train),
                              batch_size=batch_size, shuffle=True,
                              collate_fn=collate_fn)
    val_loader = DataLoader(VarSeqDataset(X_val, y_val),
                            batch_size=batch_size, shuffle=False,
                            collate_fn=collate_fn)

    # ---- モデル ----
    model = LSTMVarModel(input_dim, hidden_dim, output_dim,
                         num_layers=num_layers, dropout=dropout).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=initial_lr)
    scaler = GradScaler()

    # ============================================================
    # 学習
    # ============================================================
    for ep in range(1, num_epochs+1):
        model.train()
        for Xb, yb, Lb in train_loader:
            Xb, yb, Lb = Xb.to(device), yb.to(device), Lb.to(device)
            opt.zero_grad()

            with autocast():
                pred = model(Xb, Lb)
                B, Lmax, C = pred.shape
                ar = torch.arange(Lmax, device=device)[None, :]
                mask = (ar < Lb[:, None]).unsqueeze(-1)
                se = ((pred - yb)**2) * mask
                denom = mask.sum() * C
                loss = se.sum() / denom

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            scaler.step(opt)
            scaler.update()

        if ep % 10 == 0:
            print(f"  Epoch {ep} done.")
    # ============================================================
    # 検証
    # ============================================================
    model.eval()
    preds = []
    tgts = []
    val_raw_trial_indices = []  # store mapping to raw trial index (index in raw_X)
    with torch.no_grad():
        for batch_i, (Xv, yv, Lv) in enumerate(val_loader):
            Xv, yv, Lv = Xv.to(device), yv.to(device), Lv.to(device)
            pred = model(Xv, Lv)

            pred = pred.cpu().numpy()
            yv = yv.cpu().numpy()
            Lv = Lv.cpu().numpy()

            for i in range(pred.shape[0]):
                L = int(Lv[i])
                preds.append(pred[i, :L])
                tgts.append(yv[i, :L])
                # find the corresponding raw index: val_idx lists raw indices in order
                # We will map these sequentially: keep track by count
                val_raw_trial_indices.append(int(val_idx[len(val_raw_trial_indices)]))

    # Concatenate for metric computation
    pred_concat = np.concatenate(preds, axis=0)  # (N_total, C)
    tgt_concat = np.concatenate(tgts, axis=0)    # (N_total, C)

    # per-channel NRMSE: RMSE(channel) / range(channel)
    channel_rmse = np.sqrt(((pred_concat - tgt_concat)**2).mean(axis=0))  # shape (C,)
    channel_range = tgt_concat.max(axis=0) - tgt_concat.min(axis=0)
    channel_range[channel_range == 0] = 1.0
    channel_nrmse = channel_rmse / channel_range  # shape (C,)

    fold_nrmse = float(channel_nrmse.mean())
    fold_results.append(fold_nrmse)
    per_fold_channel_nrmse.append(channel_nrmse)

    print(f"  Fold {fold+1} channel NRMSE = {channel_nrmse}")
    print(f"  Fold {fold+1} NRMSE (mean over channels) = {fold_nrmse:.6f}")

    # ============================================================
    # NRMSE 計算に使用した元データを保存（ID 情報付き）
    # ============================================================

    # raw_trial_ids maps index in raw_X -> original trial_id
    # val_idx contains indices into raw_X; so val_raw_trial_indices is list of original raw indices used in order
    rows = []
    # val_raw_trial_indices length equals number of validation samples
    for sample_i, (p_arr, t_arr) in enumerate(zip(preds, tgts)):
        L = len(p_arr)
        raw_index = int(val_idx[sample_i])  # index into raw_X
        orig_trial_id = raw_trial_ids[raw_index]  # original trial id in df
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        for t_idx in range(L):
            row = {"raw_index": raw_index, "trial_id": orig_trial_id, "ID": subject_id, "time_idx": t_idx}
            for ch in range(output_dim):
                row[f"pred_{output_keys[ch]}"] = float(p_arr[t_idx, ch])
                row[f"tgt_{output_keys[ch]}"] = float(t_arr[t_idx, ch])
            rows.append(row)

    df_nrmse = pd.DataFrame(rows)
    df_nrmse.to_csv(f"{save_dir}/fold_{fold+1}_nrmse_data.csv", index=False)

    # ============================================================
    # 各検証トライアルの CSV 保存（ファイル名に raw trial id と subject ID を含める）
    # ============================================================
    # preds/tgts order corresponds to order of val_idx
    for i, (p, t) in enumerate(zip(preds, tgts)):
        raw_index = int(val_idx[i])
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        df_out = pd.DataFrame()
        df_out["time_index"] = np.arange(len(p))
        for ch in range(output_dim):
            df_out[f"pred_{output_keys[ch]}"] = p[:, ch]
            df_out[f"tgt_{output_keys[ch]}"] = t[:, ch]

        fname = f"{save_dir}/fold_{fold+1}_rawidx_{raw_index}_trialid_{orig_trial_id}_ID_{subject_id}.csv"
        df_out.to_csv(fname, index=False)

    # ============================================================
    # グラフ保存（ファイル名に raw trial id を含める）
    # ============================================================
    for i, (p, t) in enumerate(zip(preds, tgts)):
        raw_index = int(val_idx[i])
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        plt.figure(figsize=(12,5))
        for ch in range(p.shape[1]):
            plt.plot(t[:, ch], "--", label=f"tgt {output_keys[ch]}")
            plt.plot(p[:, ch], label=f"pred {output_keys[ch]}")
        plt.title(f"Fold {fold+1} : Validation Trial {i} (rawidx {raw_index}, trial {orig_trial_id}, ID {subject_id})")
        plt.legend()
        plt.tight_layout()
        plt.savefig(f"{fig_dir}/fold_{fold+1}_rawidx_{raw_index}_trialid_{orig_trial_id}_ID_{subject_id}.png")
        plt.close()

# ============================================================
# クロスバリデーション結果の保存
# ============================================================
cv_df = pd.DataFrame({
    "fold": list(range(1, len(fold_results)+1)),
    "NRMSE_mean_over_channels": fold_results
})
cv_df.loc[len(cv_df)] = ["mean", np.mean(fold_results)]
cv_df.to_csv(f"{save_dir}/cv_results.csv", index=False)

# Optionally save per-fold per-channel NRMSE table
channel_cols = [f"NRMSE_{k}" for k in output_keys]
channel_table = pd.DataFrame(np.vstack(per_fold_channel_nrmse), columns=channel_cols)
channel_table.insert(0, "fold", list(range(1, len(per_fold_channel_nrmse)+1)))
channel_table.loc[len(channel_table)] = ["mean"] + list(channel_table[channel_cols].mean(axis=0))
channel_table.to_csv(f"{save_dir}/cv_channel_nrmse.csv", index=False)

print("="*60)
print(" 5Fold CV 結果")
print("="*60)
for i, v in enumerate(fold_results):
    print(f" Fold {i+1} NRMSE = {v:.6f}")
print(" 平均 NRMSE =", np.mean(fold_results))
print(f"\n→ 保存先: {save_dir}/cv_results.csv")
```

```python
# ============================================================
# improved_gait_transformer_variable_length_5fold_cv_fixed_with_earlystop.py
# Transformer 版: 可変長 Transformer Encoder + Δt 分割 + 5Fold CV + NRMSE + EarlyStopping
# ============================================================

import os
import random
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

from sklearn.model_selection import KFold
from sklearn.model_selection import GroupKFold

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
from torch.nn.utils.rnn import pad_sequence

# ============================================================
# 0. reproducibility
# ============================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ============================================================
# EarlyStopping クラス
# ============================================================
class EarlyStopping:
    def __init__(self, patience=10, delta=1e-5):
        self.patience = patience
        self.delta = delta
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.best_state = None

    def step(self, score, model):
        """
        score: validation RMSE（小さいほど良い）
        """
        if self.best_score is None:
            self.best_score = score
            self.best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            return False

        if score < self.best_score - self.delta:
            # 改善
            self.best_score = score
            self.best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            self.counter = 0
            return False
        else:
            # 改善なし
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
                return True
            return False

# ============================================================
# 1. 設定
# ============================================================

file_path = "/content/new_wf_tr_extraction_selected.csv"
encoding = "cp932"
low_memory = False

batch_size = 8
num_epochs = 1000
initial_lr = 5e-4

hidden_dim = 96
num_layers = 2
dropout = 0.2
grad_clip = 1.0
nhead = 4

input_keys = [ "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm",
"左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm",
"左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm",
"左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm",
"左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm",
"左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm",
"右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm",
"右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm",
"右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm",
"右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm",
"右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm",
"右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm",
"左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm",
"左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm",
"左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm",
"左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm",
"左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm",
"左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm",
"右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm",
"右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm",
"右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm",
"右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm",
"PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm",
"股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm",
"肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm",
"体重心Xcm","体重心Ycm","体重心Zcm" ]

output_keys = ["左足Fzkg","左足Fxkg","左足Fykg","右足Fzkg","右足Fxkg","右足Fykg"]

# 保存ディレクトリ
save_dir = "/content/drive/MyDrive/t-teramae/results"
os.makedirs(save_dir, exist_ok=True)
fig_dir = "/content/drive/MyDrive/t-teramae/fig"
os.makedirs(fig_dir, exist_ok=True)

# ============================================================
# 2. CSV 読み込み
# ============================================================
df = pd.read_csv(file_path, encoding=encoding, low_memory=low_memory)
print("CSV 読み込み完了:", len(df))

# 時間列検出
possible_time_cols = ["time", "Time", "timestamp", "時間", "TIME"]
TIME_COL = next((c for c in possible_time_cols if c in df.columns), None)
if TIME_COL is None:
    raise KeyError("時間列が見つかりません")

df[TIME_COL] = df[TIME_COL].astype(float)

# ============================================================
# 3. Δt 分割
# ============================================================
tol = 5e-4
dt = df[TIME_COL].diff().fillna(0.0).values
is_regular = np.isclose(dt, 0.001, atol=tol)

trial_id = []
cur = 0
for i, reg in enumerate(is_regular):
    if i == 0:
        trial_id.append(cur)
        continue
    if not reg:
        cur += 1
    trial_id.append(cur)

df["trial_id"] = trial_id
num_trials = int(df["trial_id"].max() + 1)
print("検出されたトライアル:", num_trials)

# ============================================================
# 4. トライアル単位で抽出
# ============================================================
raw_X, raw_y, raw_trial_ids = [], [], []
MIN_LEN = 5

for tid in range(num_trials):
    chunk = df[df["trial_id"] == tid]
    if len(chunk) < MIN_LEN:
        continue
    if chunk[input_keys].isna().any().any() or chunk[output_keys].isna().any().any():
        continue

    raw_X.append(chunk[input_keys].values.astype(np.float32))
    raw_y.append(chunk[output_keys].values.astype(np.float32))
    raw_trial_ids.append(int(tid))

print("有効トライアル:", len(raw_X))

if len(raw_X) == 0:
    raise RuntimeError("有効なトライアルが0件です")

MAX_LEN = max([x.shape[0] for x in raw_X])
print("最大系列長:", MAX_LEN)

# ============================================================
# Dataset
# ============================================================
class VarSeqDataset(Dataset):
    def __init__(self, X_list, y_list, idx_list):
        self.X = X_list
        self.y = y_list
        self.idx = idx_list

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return (
            torch.tensor(self.X[idx]),
            torch.tensor(self.y[idx]),
            int(self.X[idx].shape[0]),
            int(self.idx[idx])
        )

def collate_fn(batch):
    # batch: list of (x, y, length, raw_index)
    xs, ys, lengths, raw_indices = zip(*batch)
    xs_pad = pad_sequence(xs, batch_first=True)  # (B, Lmax, D_in)
    ys_pad = pad_sequence(ys, batch_first=True)  # (B, Lmax, D_out)
    lengths = torch.tensor(lengths, dtype=torch.long)  # 明示的に long
    raw_indices = torch.tensor(raw_indices, dtype=torch.long)
    return xs_pad, ys_pad, lengths, raw_indices

# ============================================================
# Transformer モデル
# ============================================================
class TransformerVarModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim,
                 num_layers=4, dropout=0.1, nhead=4, max_len=1024):
        super().__init__()
        self.input_fc = nn.Linear(input_dim, hidden_dim)

        # learnable positional embedding（MAX_LEN を渡すこと）
        self.pos_embed = nn.Parameter(torch.randn(1, max_len, hidden_dim))

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim,
            nhead=nhead,
            dim_feedforward=hidden_dim * 4,
            dropout=dropout,
            batch_first=True
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        self.drop = nn.Dropout(dropout)
        self.fc_out = nn.Linear(hidden_dim, output_dim)

    def forward(self, x_pad, lengths):
        """
        x_pad: (B, Lmax, input_dim)
        lengths: (B,)
        """
        B, Lmax, _ = x_pad.shape
        x = self.input_fc(x_pad)  # (B, Lmax, H)

        if self.pos_embed.size(1) < Lmax:
            raise ValueError(f"pos_embed max_len ({self.pos_embed.size(1)}) < Lmax ({Lmax}). Increase max_len.")
        x = x + self.pos_embed[:, :Lmax, :]

        # padding mask: True where padding (to be masked)
        pad_mask = torch.arange(Lmax, device=lengths.device)[None, :] >= lengths[:, None]  # (B, Lmax)
        out = self.encoder(x, src_key_padding_mask=pad_mask)

        out = self.drop(out)
        out = self.fc_out(out)  # (B, Lmax, output_dim)
        return out

# ============================================================
# 5-fold GroupKFold CV + 学習ループ（EarlyStopping 統合）
# ============================================================
fold_results = []
per_fold_channel_nrmse = []

# trial -> subject mapping
trial_to_subject = df.groupby("trial_id")["ID"].first().to_dict()

# raw_X の各要素の subject ID を groups とする
groups = np.array([trial_to_subject[raw_trial_ids[i]] for i in range(len(raw_X))])

gkf = GroupKFold(n_splits=5)

for fold, (train_idx, val_idx) in enumerate(gkf.split(raw_X, groups=groups)):
    print()
    print("="*60)
    print(f" Fold {fold+1} / 5  (GroupKFold)")
    print("="*60)

    # データ分割
    X_train = [raw_X[i] for i in train_idx]
    y_train = [raw_y[i] for i in train_idx]
    idx_train = [i for i in train_idx]

    X_val   = [raw_X[i] for i in val_idx]
    y_val   = [raw_y[i] for i in val_idx]
    idx_val = [i for i in val_idx]

    # 標準化（train の concat）
    concat_Xtrain = np.concatenate(X_train, axis=0)
    mean = concat_Xtrain.mean(axis=0)
    std = concat_Xtrain.std(axis=0)
    std[std == 0] = 1.0

    X_train = [(x - mean)/std for x in X_train]
    X_val = [(x - mean)/std for x in X_val]

    train_dataset = VarSeqDataset(X_train, y_train, idx_train)
    val_dataset = VarSeqDataset(X_val, y_val, idx_val)

    train_loader = DataLoader(train_dataset,
                              batch_size=batch_size, shuffle=True,
                              collate_fn=collate_fn, drop_last=False, pin_memory=True)
    val_loader = DataLoader(val_dataset,
                            batch_size=batch_size, shuffle=False,
                            collate_fn=collate_fn, drop_last=False, pin_memory=True)

    input_dim = raw_X[0].shape[1]
    output_dim = raw_y[0].shape[1]

    model = TransformerVarModel(
        input_dim=input_dim,
        hidden_dim=hidden_dim,
        output_dim=output_dim,
        num_layers=num_layers,
        dropout=dropout,
        nhead=nhead,
        max_len=MAX_LEN
    ).to(device)

    opt = torch.optim.Adam(model.parameters(), lr=initial_lr)
    scaler = GradScaler()

    # EarlyStopping（fold ごと）
    earlystop = EarlyStopping(patience=10, delta=1e-5)

    # 学習ループ（各 epoch ごとに validation を実行し、EarlyStopping 判定）
    for ep in range(1, num_epochs+1):
        model.train()
        epoch_loss = 0.0
        count_batches = 0

        for Xb, yb, Lb, raw_idx_batch in train_loader:
            Xb, yb, Lb = Xb.to(device), yb.to(device), Lb.to(device)
            opt.zero_grad()

            # マスク付き損失（安定実装）
            with autocast():
                pred = model(Xb, Lb)   # (B, Lmax, C_out)

                B, Lmax, C = pred.shape
                ar = torch.arange(Lmax, device=Lb.device)[None, :]
                mask = (ar < Lb[:, None]).unsqueeze(-1)  # (B, Lmax, 1), bool

                y_true = yb.to(dtype=pred.dtype)
                se = ((pred - y_true) ** 2) * mask
                se_sum = se.sum()
                valid_count = mask.sum().item() * C

                if valid_count == 0:
                    loss = se_sum * 0.0
                else:
                    loss = se_sum / float(valid_count)

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            scaler.step(opt)
            scaler.update()

            epoch_loss += float(loss.detach().cpu().item())
            count_batches += 1

        avg_train_loss = epoch_loss / max(1, count_batches)

        # ---------- validation（epoch 内） ----------
        model.eval()
        val_preds = []
        val_tgts = []
        with torch.no_grad():
            for Xv, yv, Lv, _ in val_loader:
                Xv, yv, Lv = Xv.to(device), yv.to(device), Lv.to(device)
                pv = model(Xv, Lv)  # (B, Lmax, C_out)

                pv_np = pv.cpu().numpy()
                yv_np = yv.cpu().numpy()
                Lv_np = Lv.cpu().numpy()

                for i in range(pv_np.shape[0]):
                    L = int(Lv_np[i])
                    val_preds.append(pv_np[i, :L])
                    val_tgts.append(yv_np[i, :L])

        if len(val_preds) == 0:
            print(f"  Epoch {ep}: validation データがありません（fold {fold+1}）")
            val_score = float("inf")
        else:
            pred_concat = np.concatenate(val_preds, axis=0)
            tgt_concat = np.concatenate(val_tgts, axis=0)

            # 全 channel の RMSE（earlystop 指標）
            val_rmse = np.sqrt(((pred_concat - tgt_concat)**2).mean())
            val_score = float(val_rmse)

        print(f"  Epoch {ep:3d}  TrainLoss={avg_train_loss:.6f}  ValRMSE={val_score if np.isfinite(val_score) else 'inf'}")

        # EarlyStopping 判定（小さい方が良い指標）
        stop = earlystop.step(val_score, model)
        if stop:
            print(f"  EarlyStopping triggered at epoch {ep} (fold {fold+1})")
            break

    # 学習終了後、ベストモデルで復元（EarlyStopping が保存した best_state）
    if earlystop.best_state is not None:
        model.load_state_dict(earlystop.best_state)
        print(f"  Best model restored for fold {fold+1} (val_score={earlystop.best_score})")

    # ============================================================
    # fold ごとの検証（最終復元モデルで詳細評価 & 保存）
    # （ここは元コードの検証 / CSV / plot 保存処理に合わせています）
    # ============================================================
    model.eval()
    preds = []
    tgts = []
    val_raw_indices_for_rows = []
    with torch.no_grad():
        for batch_i, (Xv, yv, Lv, raw_idx_batch) in enumerate(val_loader):
            Xv, yv, Lv = Xv.to(device), yv.to(device), Lv.to(device)
            pred = model(Xv, Lv)  # (B, Lmax, C_out)

            pred = pred.cpu().numpy()
            yv_np = yv.cpu().numpy()
            Lv_np = Lv.cpu().numpy()
            raw_idx_np = raw_idx_batch.numpy()

            for i in range(pred.shape[0]):
                L = int(Lv_np[i])
                preds.append(pred[i, :L])
                tgts.append(yv_np[i, :L])
                val_raw_indices_for_rows.append(int(raw_idx_np[i]))

    # guard
    if len(preds) == 0:
        print(f"  Fold {fold+1}: 検証結果が空です。スキップします。")
        fold_results.append(float("nan"))
        per_fold_channel_nrmse.append(np.array([float("nan")] * output_dim))

        # モデルは保存しておく
        torch.save({
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": opt.state_dict(),
            "mean": mean,
            "std": std,
            "fold": fold+1,
            "input_keys": input_keys,
            "output_keys": output_keys
        }, f"{save_dir}/fold_{fold+1}_model.pt")
        continue

    pred_concat = np.concatenate(preds, axis=0)  # (N_total, C)
    tgt_concat = np.concatenate(tgts, axis=0)    # (N_total, C)

    channel_rmse = np.sqrt(((pred_concat - tgt_concat)**2).mean(axis=0))
    channel_range = tgt_concat.max(axis=0) - tgt_concat.min(axis=0)
    channel_range[channel_range == 0] = 1.0
    channel_nrmse = channel_rmse / channel_range

    fold_nrmse = float(channel_nrmse.mean())
    fold_results.append(fold_nrmse)
    per_fold_channel_nrmse.append(channel_nrmse)

    print(f"  Fold {fold+1} channel NRMSE = {channel_nrmse}")
    print(f"  Fold {fold+1} NRMSE (mean over channels) = {fold_nrmse:.6f}")

    # NRMSE 用の元データ保存（ID 情報付き）
    rows = []
    for sample_i, (p_arr, t_arr, raw_index) in enumerate(zip(preds, tgts, val_raw_indices_for_rows)):
        L = len(p_arr)
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        for t_idx in range(L):
            row = {"raw_index": raw_index, "trial_id": orig_trial_id, "ID": subject_id, "time_idx": t_idx}
            for ch in range(output_dim):
                row[f"pred_{output_keys[ch]}"] = float(p_arr[t_idx, ch])
                row[f"tgt_{output_keys[ch]}"] = float(t_arr[t_idx, ch])
            rows.append(row)

    df_nrmse = pd.DataFrame(rows)
    df_nrmse.to_csv(f"{save_dir}/fold_{fold+1}_nrmse_data.csv", index=False)

    # 各検証トライアルの保存（CSV とプロット）
    for i, (p, t, raw_index) in enumerate(zip(preds, tgts, val_raw_indices_for_rows)):
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        df_out = pd.DataFrame()
        df_out["time_index"] = np.arange(len(p))
        for ch in range(output_dim):
            df_out[f"pred_{output_keys[ch]}"] = p[:, ch]
            df_out[f"tgt_{output_keys[ch]}"] = t[:, ch]

        fname = f"{save_dir}/fold_{fold+1}_rawidx_{raw_index}_trialid_{orig_trial_id}_ID_{subject_id}.csv"
        df_out.to_csv(fname, index=False)

    for i, (p, t, raw_index) in enumerate(zip(preds, tgts, val_raw_indices_for_rows)):
        orig_trial_id = raw_trial_ids[raw_index]
        subject_id = trial_to_subject.get(orig_trial_id, -1)

        plt.figure(figsize=(12,5))
        for ch in range(p.shape[1]):
            plt.plot(t[:, ch], "--", label=f"tgt {output_keys[ch]}")
            plt.plot(p[:, ch], label=f"pred {output_keys[ch]}")
        plt.title(f"Fold {fold+1} : Validation Trial {i} (rawidx {raw_index}, trial {orig_trial_id}, ID {subject_id})")
        plt.legend()
        plt.tight_layout()
        plt.savefig(f"{fig_dir}/fold_{fold+1}_rawidx_{raw_index}_trialid_{orig_trial_id}_ID_{subject_id}.png")
        plt.close()

    # fold モデルの保存
    torch.save({
        "model_state_dict": model.state_dict(),
        "optimizer_state_dict": opt.state_dict(),
        "mean": mean,
        "std": std,
        "fold": fold+1,
        "input_keys": input_keys,
        "output_keys": output_keys
    }, f"{save_dir}/fold_{fold+1}_model.pt")

# ============================================================
# CV 結果の保存
# ============================================================
cv_df = pd.DataFrame({
    "fold": list(range(1, len(fold_results)+1)),
    "NRMSE_mean_over_channels": fold_results
})
cv_df.loc[len(cv_df)] = ["mean", np.nanmean(fold_results)]
cv_df.to_csv(f"{save_dir}/cv_results.csv", index=False)

channel_cols = [f"NRMSE_{k}" for k in output_keys]
if len(per_fold_channel_nrmse) > 0:
    channel_table = pd.DataFrame(np.vstack([np.array(x) for x in per_fold_channel_nrmse]), columns=channel_cols)
    channel_table.insert(0, "fold", list(range(1, len(per_fold_channel_nrmse)+1)))
    mean_row = ["mean"] + list(channel_table[channel_cols].mean(axis=0))
    channel_table.loc[len(channel_table)] = mean_row
    channel_table.to_csv(f"{save_dir}/cv_channel_nrmse.csv", index=False)
else:
    pd.DataFrame(columns=["fold"] + channel_cols).to_csv(f"{save_dir}/cv_channel_nrmse.csv", index=False)

print("="*60)
print(" 5Fold CV 結果")
print("="*60)
for i, v in enumerate(fold_results):
    print(f" Fold {i+1} NRMSE = {v}")
print(" 平均 NRMSE =", np.nanmean(fold_results))
print(f"\n→ 保存先: {save_dir}/cv_results.csv")
```

```python
# ============================================================
# train_and_save_for_paper.py
# ============================================================

import os, random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupKFold
from collections import defaultdict

# ============================================================
# 1. reproducibility
# ============================================================
SEED = 30
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================================================
# 2. experiment setting
# ============================================================
MODEL_TYPE = "LSTM"   # "TCN" or "LSTM"

EXP_DIR = f"experiment_fair_{MODEL_TYPE.lower()}"
for sub in ["models","predictions","metrics"]:
    os.makedirs(f"{EXP_DIR}/{sub}", exist_ok=True)

file_path = "new_wf_tr_extraction_selected.csv"
encoding = "cp932"

WINDOW = 1000
STRIDE = 250
batch_size = 8
num_epochs = 1000
patience = 50
weight_decay = 1e-5

# ---- Fair capacity ----
hidden_ch = 32
levels = 7
kernel_size = 5
dropout = 0.1

lstm_hidden = 256
lstm_layers = 3

lr = 1e-4 if MODEL_TYPE=="TCN" else 3e-4

# ============================================================
# 3. keys
# ============================================================
input_keys = [ "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm",
"左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm",
"左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm",
"左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm",
"左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm",
"左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm",
"右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm",
"右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm",
"右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm",
"右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm",
"右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm",
"右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm",
"左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm",
"左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm",
"左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm",
"左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm",
"左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm",
"左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm",
"右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm",
"右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm",
"右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm",
"右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm",
"PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm",
"股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm",
"肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm",
"体重心Xcm","体重心Ycm","体重心Zcm" ]

output_keys = ["左足Fzkg","左足Fxkg","左足Fykg","右足Fzkg","右足Fxkg","右足Fykg"]
WEIGHT_COL = "体重"

# ============================================================
# 4. load & windowing
# ============================================================
df = pd.read_csv(file_path, encoding=encoding)
TIME_COL = next(c for c in ["time","Time","timestamp","時間"] if c in df.columns)
df[TIME_COL] = df[TIME_COL].astype(float)

df["trial_id"] = (
    (df["ID"] != df["ID"].shift()) |
    (df[TIME_COL].diff() > 0.005)
).cumsum()

trial_len = df.groupby("trial_id").size()
df = df[df["trial_id"].isin(trial_len[trial_len>=WINDOW].index)].reset_index(drop=True)

X, y, groups, trial_ids, window_starts = [], [], [], [], []

for tid, g in df.groupby("trial_id"):
    if g[input_keys + output_keys + [WEIGHT_COL]].isna().any().any():
        continue

    Xv = g[input_keys].values.astype(np.float32)
    yv = g[output_keys].values.astype(np.float32) / g[WEIGHT_COL].iloc[0] * 100

    for s in range(0, len(Xv)-WINDOW+1, STRIDE):
        X.append(Xv[s:s+WINDOW])
        y.append(yv[s:s+WINDOW])
        groups.append(g["ID"].iloc[0])
        trial_ids.append(tid)
        window_starts.append(s)

X, y = np.stack(X), np.stack(y)

# ============================================================
# 5. Dataset
# ============================================================
class WindowDataset(Dataset):
    def __init__(self, X, y, tid, ws):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
        self.tid = tid
        self.ws = ws
    def __len__(self): return len(self.X)
    def __getitem__(self, i):
        return self.X[i], self.y[i], self.tid[i], self.ws[i]

# ============================================================
# 6. Models
# ============================================================
class TemporalBlock(nn.Module):
    def __init__(self, in_ch, out_ch, k, d, drop):
        super().__init__()
        p = (k-1)*d
        self.c1 = nn.Conv1d(in_ch,out_ch,k,padding=p,dilation=d)
        self.c2 = nn.Conv1d(out_ch,out_ch,k,padding=p,dilation=d)
        self.do = nn.Dropout(drop)
        self.ds = nn.Conv1d(in_ch,out_ch,1) if in_ch!=out_ch else None

    def forward(self,x):
        y = self.do(F.relu(self.c1(x)[:,:,:-self.c1.padding[0]]))
        y = self.do(F.relu(self.c2(y)[:,:,:-self.c2.padding[0]]))
        r = x if self.ds is None else self.ds(x)
        return F.relu(y+r)

class TCN(nn.Module):
    def __init__(self, inp, out):
        super().__init__()
        ch = inp
        blocks=[]
        for i in range(levels):
            blocks.append(TemporalBlock(ch,hidden_ch,kernel_size,2**i,dropout))
            ch = hidden_ch
        self.net = nn.Sequential(*blocks)
        self.fc = nn.Conv1d(hidden_ch,out,1)

    def forward(self,x):
        return self.fc(self.net(x.transpose(1,2))).transpose(1,2)

class LSTMModel(nn.Module):
    def __init__(self, inp, out):
        super().__init__()
        self.lstm = nn.LSTM(inp,lstm_hidden,lstm_layers,
                            batch_first=True,dropout=dropout)
        self.fc = nn.Linear(lstm_hidden,out)

    def forward(self,x):
        y,_ = self.lstm(x)
        return self.fc(y)

# ============================================================
# 7. trial-wise NRMSE + reconstruction
# ============================================================
def trialwise_eval(model, loader, ym, ys):
    model.eval()
    P,T,S,L = defaultdict(list),defaultdict(list),defaultdict(list),defaultdict(int)

    with torch.no_grad():
        for xb,yb,tid,ws in loader:
            pr = model(xb.to(device)).cpu().numpy()*ys + ym
            tr = yb.numpy()*ys + ym
            for i in range(len(pr)):
                t,s = int(tid[i]), int(ws[i])
                P[t].append(pr[i]); T[t].append(tr[i]); S[t].append(s)
                L[t] = max(L[t], s+WINDOW)

    trial_scores, recon = {}, {}

    for t in P:
        ps = np.zeros((L[t],y.shape[-1]))
        ts = np.zeros_like(ps)
        c  = np.zeros((L[t],1))
        for p,tr,s in zip(P[t],T[t],S[t]):
            ps[s:s+WINDOW]+=p; ts[s:s+WINDOW]+=tr; c[s:s+WINDOW]+=1
        v = c.squeeze()>0
        ps, ts = ps[v]/c[v], ts[v]/c[v]

        rmse = np.sqrt(np.mean((ps-ts)**2,0))
        rng = np.ptp(ts,0); rng[rng==0]=1
        trial_scores[t] = (rmse/rng*100).mean()
        recon[t] = (ps, ts)

    return np.mean(list(trial_scores.values())), trial_scores, recon

# ============================================================
# 8. Cross-validation
# ============================================================
gkf = GroupKFold(5)
scores = []

for f,(tr,va) in enumerate(gkf.split(X,y,groups)):
    Xm,Xs = X[tr].mean((0,1)),X[tr].std((0,1))
    Ym,Ys = y[tr].mean((0,1)),y[tr].std((0,1))
    Xs[Xs==0]=1; Ys[Ys==0]=1

    tr_ds = WindowDataset((X[tr]-Xm)/Xs,(y[tr]-Ym)/Ys,
                          np.array(trial_ids)[tr],np.array(window_starts)[tr])
    va_ds = WindowDataset((X[va]-Xm)/Xs,(y[va]-Ym)/Ys,
                          np.array(trial_ids)[va],np.array(window_starts)[va])

    model = TCN(X.shape[-1],y.shape[-1]) if MODEL_TYPE=="TCN" else LSTMModel(X.shape[-1],y.shape[-1])
    model.to(device)

    opt = torch.optim.AdamW(model.parameters(),lr=lr,weight_decay=weight_decay)

    best, wait = 1e9, 0

    for ep in range(num_epochs):
        model.train()
        for xb,yb,_,_ in DataLoader(tr_ds,batch_size,True):
            opt.zero_grad()
            loss = F.mse_loss(model(xb.to(device)), yb.to(device))
            loss.backward()
            if MODEL_TYPE=="LSTM":
                torch.nn.utils.clip_grad_norm_(model.parameters(),1.0)
            opt.step()

        sc, trial_scores, recon = trialwise_eval(
            model,
            DataLoader(va_ds,batch_size,False),
            Ym, Ys
        )

        if sc < best:
            best, wait = sc, 0

            # model
            torch.save(model.state_dict(), f"{EXP_DIR}/models/fold{f}.pt")

            # predictions
            P_all, T_all, tid_all = [], [], []
            for t,(p,trg) in recon.items():
                P_all.append(p); T_all.append(trg)
                tid_all.extend([t]*len(p))

            np.savez(
                f"{EXP_DIR}/predictions/fold{f}.npz",
                pred=np.vstack(P_all),
                gt=np.vstack(T_all),
                trial_id=np.array(tid_all)
            )

            pd.DataFrame({
                "trial_id": list(trial_scores.keys()),
                "nrmse": list(trial_scores.values())
            }).to_csv(
                f"{EXP_DIR}/metrics/fold{f}_trial_nrmse.csv",
                index=False
            )

        else:
            wait += 1
            if wait >= patience:
                break

    scores.append(best)

# ============================================================
# 9. CV summary & config
# ============================================================
pd.DataFrame({
    "fold": np.arange(len(scores)),
    "nrmse": scores,
    "model": MODEL_TYPE
}).to_csv(f"{EXP_DIR}/metrics/cv_summary.csv", index=False)

with open(f"{EXP_DIR}/config.txt","w") as f:
    f.write(f"MODEL_TYPE={MODEL_TYPE}\n")
    f.write(f"WINDOW={WINDOW}, STRIDE={STRIDE}\n")
    f.write(f"SEED={SEED}\n")

print("Training & saving completed.")
print("Mean NRMSE:", np.mean(scores))
```

```python
# ============================================================
# analysis_for_paper.py
# ============================================================
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import wilcoxon

sns.set(style="whitegrid", font_scale=1.3)

# ============================================================
# 1. load trial-wise NRMSE
# ============================================================
def load_trial_nrmse(exp_dir):
    dfs = []
    for f in os.listdir(f"{exp_dir}/metrics"):
        if f.endswith("_trial_nrmse.csv"):
            df = pd.read_csv(f"{exp_dir}/metrics/{f}")
            dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

tcn_dir  = "experiment_fair_tcn"
lstm_dir = "experiment_fair_lstm"

tcn_df  = load_trial_nrmse(tcn_dir)
lstm_df = load_trial_nrmse(lstm_dir)

tcn_df["model"]  = "TCN"
lstm_df["model"] = "LSTM"

# ============================================================
# 2. paired dataset（trial単位で整列）
# ============================================================
merged = pd.merge(
    tcn_df, lstm_df,
    on="trial_id",
    suffixes=("_tcn", "_lstm")
)

# ============================================================
# 3. 統計検定（paired Wilcoxon）
# ============================================================
stat, p_value = wilcoxon(
    merged["nrmse_tcn"],
    merged["nrmse_lstm"]
)

print("Paired Wilcoxon test")
print(f"  statistic = {stat:.3f}")
print(f"  p-value   = {p_value:.5f}")

# effect size (rank-biserial)
diff = merged["nrmse_tcn"] - merged["nrmse_lstm"]
rbc = (np.sum(diff > 0) - np.sum(diff < 0)) / len(diff)
print(f"  rank-biserial correlation = {rbc:.3f}")

# ============================================================
# 4. Table（論文用 Mean ± SD）
# ============================================================
table = pd.DataFrame({
    "Model": ["TCN", "LSTM"],
    "Mean NRMSE (%)": [
        merged["nrmse_tcn"].mean(),
        merged["nrmse_lstm"].mean()
    ],
    "SD": [
        merged["nrmse_tcn"].std(),
        merged["nrmse_lstm"].std()
    ]
})

print("\nTable (Mean ± SD)")
print(table)

table.to_csv("table_nrmse_summary.csv", index=False)

# ============================================================
# 5. Figure 1: Boxplot (TCN vs LSTM)
# ============================================================
plot_df = pd.DataFrame({
    "NRMSE (%)": np.concatenate([merged["nrmse_tcn"], merged["nrmse_lstm"]]),
    "Model": ["TCN"]*len(merged) + ["LSTM"]*len(merged)
})

plt.figure(figsize=(6,6))
sns.boxplot(data=plot_df, x="Model", y="NRMSE (%)", showfliers=False)
sns.stripplot(data=plot_df, x="Model", y="NRMSE (%)",
              color="black", alpha=0.5, jitter=True)

plt.title("Trial-wise NRMSE Comparison")
plt.tight_layout()
plt.savefig("fig_boxplot_nrmse.png", dpi=300)
plt.close()

# ============================================================
# 6. Figure 2: Paired line plot
# ============================================================
plt.figure(figsize=(6,6))
for _, r in merged.iterrows():
    plt.plot(["TCN","LSTM"], [r["nrmse_tcn"], r["nrmse_lstm"]],
             color="gray", alpha=0.4)

plt.scatter(["TCN"]*len(merged), merged["nrmse_tcn"], label="TCN")
plt.scatter(["LSTM"]*len(merged), merged["nrmse_lstm"], label="LSTM")

plt.ylabel("NRMSE (%)")
plt.title("Paired Trial-wise Comparison")
plt.tight_layout()
plt.savefig("fig_paired_nrmse.png", dpi=300)
plt.close()

# ============================================================
# 7. Figure 3: Representative GRF waveform
# ============================================================
def load_prediction(exp_dir, fold=0):
    d = np.load(f"{exp_dir}/predictions/fold{fold}.npz")
    return d["pred"], d["gt"], d["trial_id"]

# trialを1つ選ぶ（中央値）
trial_id = merged["trial_id"].iloc[len(merged)//2]

p_tcn, g_tcn, tid_tcn = load_prediction(tcn_dir)
p_lstm, g_lstm, tid_lstm = load_prediction(lstm_dir)

mask_tcn  = tid_tcn == trial_id
mask_lstm = tid_lstm == trial_id

plt.figure(figsize=(10,4))
plt.plot(g_tcn[mask_tcn][:,0], label="GT", color="black", linewidth=2)
plt.plot(p_tcn[mask_tcn][:,0], label="TCN", linestyle="--")
plt.plot(p_lstm[mask_lstm][:,0], label="LSTM", linestyle=":")

plt.xlabel("Time frame")
plt.ylabel("GRF (%BW)")
plt.title(f"Representative Trial (ID={trial_id})")
plt.legend()
plt.tight_layout()
plt.savefig("fig_waveform_example.png", dpi=300)
plt.close()

print("\nAll figures and tables saved successfully.")
```

```python
# ============================================================
# 0. imports
# ============================================================
import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupKFold
from collections import defaultdict

# ============================================================
# 1. reproducibility
# ============================================================
SEED = 30
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ============================================================
# 2. experiment setting
# ============================================================
EXP_DIR = "/content/drive/MyDrive/t-teramae/experiment_tcn_mc_dropout"
os.makedirs(EXP_DIR, exist_ok=True)

file_path = "/content/new_wf_tr_extraction_selected.csv"
encoding = "cp932"

WINDOW = 1000
STRIDE = 250
batch_size = 8
num_epochs = 1000
lr = 1e-4

hidden_ch = 64
levels = 5
kernel_size = 5
dropout = 0.2          # ★ MC Dropout は少し強めが良い
weight_decay = 1e-5
patience = 50

MC_SAMPLES = 20        # ★ forward 回数

# ============================================================
# 3. input / output keys
# ============================================================
input_keys = [
    "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm",
    "左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm",
    "左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm",
    "左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm",
    "左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm",
    "左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm",
    "右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm",
    "右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm",
    "右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm",
    "右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm",
    "右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm",
    "右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm",
    "左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm",
    "左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm",
    "左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm",
    "左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm",
    "左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm",
    "左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm",
    "右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm",
    "右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm",
    "右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm",
    "右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm",
    "PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm",
    "股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm",
    "肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm",
    "体重心Xcm","体重心Ycm","体重心Zcm"
]
output_keys = [
    "左足Fzkg","左足Fxkg","左足Fykg",
    "右足Fzkg","右足Fxkg","右足Fykg"
]
WEIGHT_COL = "体重"

# ============================================================
# 4. load & windowing
# ============================================================
df = pd.read_csv(file_path, encoding=encoding)
TIME_COL = next(c for c in ["time","Time","timestamp","時間"] if c in df.columns)
df[TIME_COL] = df[TIME_COL].astype(float)

df["trial_id"] = (
    (df["ID"] != df["ID"].shift()) |
    (df[TIME_COL].diff() > 0.005)
).cumsum()

trial_len = df.groupby("trial_id").size()
valid_trials = trial_len[trial_len >= WINDOW].index
df = df[df["trial_id"].isin(valid_trials)].reset_index(drop=True)

X, y, groups, trial_ids, window_starts = [], [], [], [], []

for tid, g in df.groupby("trial_id"):
    if g[input_keys + output_keys + [WEIGHT_COL]].isna().any().any():
        continue

    Xv = g[input_keys].values.astype(np.float32)
    yv = g[output_keys].values.astype(np.float32) / g[WEIGHT_COL].iloc[0] * 100.0

    for s in range(0, len(Xv)-WINDOW+1, STRIDE):
        X.append(Xv[s:s+WINDOW])
        y.append(yv[s:s+WINDOW])
        groups.append(g["ID"].iloc[0])
        trial_ids.append(tid)
        window_starts.append(s)

X = np.stack(X)
y = np.stack(y)

# ============================================================
# 5. Dataset
# ============================================================
class WindowDataset(Dataset):
    def __init__(self, X, y, tid, wstart):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
        self.tid = tid
        self.wstart = wstart

    def __len__(self): return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i], self.tid[i], self.wstart[i]

# ============================================================
# 6. TCN with Dropout
# ============================================================
class TemporalBlock(nn.Module):
    def __init__(self, in_ch, out_ch, k, d, drop):
        super().__init__()
        p = (k - 1) * d
        self.c1 = nn.Conv1d(in_ch, out_ch, k, padding=p, dilation=d)
        self.c2 = nn.Conv1d(out_ch, out_ch, k, padding=p, dilation=d)
        self.do = nn.Dropout(drop)
        self.ds = nn.Conv1d(in_ch, out_ch, 1) if in_ch != out_ch else None

    def forward(self, x):
        y = self.do(F.relu(self.c1(x)[:, :, :-self.c1.padding[0]]))
        y = self.do(F.relu(self.c2(y)[:, :, :-self.c2.padding[0]]))
        r = x if self.ds is None else self.ds(x)
        return F.relu(y + r)

class TCN(nn.Module):
    def __init__(self, inp, out):
        super().__init__()
        ch = inp
        blocks = []
        for i in range(levels):
            blocks.append(
                TemporalBlock(ch, hidden_ch, kernel_size, 2**i, dropout)
            )
            ch = hidden_ch
        self.net = nn.Sequential(*blocks)
        self.fc = nn.Conv1d(hidden_ch, out, 1)

    def forward(self, x):
        x = x.transpose(1, 2)
        y = self.fc(self.net(x))
        return y.transpose(1, 2)

# ============================================================
# 7. trial-wise NRMSE with MC Dropout
# ============================================================
def trialwise_nrmse_mc(model, loader, ym, ys):
    model.train()   # ★ Dropout ON
    P, T, S, L = defaultdict(list), defaultdict(list), defaultdict(list), defaultdict(int)

    with torch.no_grad():
        for xb, yb, tid, ws in loader:
            xb = xb.to(device)

            preds = []
            for _ in range(MC_SAMPLES):
                preds.append(model(xb).cpu().numpy())
            pred = np.mean(preds, axis=0) * ys + ym
            true = yb.numpy() * ys + ym

            for i in range(len(pred)):
                t = int(tid[i])
                s = int(ws[i])
                P[t].append(pred[i])
                T[t].append(true[i])
                S[t].append(s)
                L[t] = max(L[t], s + WINDOW)

    scores = []
    for t in P:
        ps = np.zeros((L[t], y.shape[-1]))
        ts = np.zeros((L[t], y.shape[-1]))
        c = np.zeros((L[t], 1))

        for p, tr, s in zip(P[t], T[t], S[t]):
            ps[s:s+WINDOW] += p
            ts[s:s+WINDOW] += tr
            c[s:s+WINDOW] += 1

        v = c.squeeze() > 0
        rmse = np.sqrt(np.mean((ps[v]/c[v] - ts[v]/c[v])**2, axis=0))
        rng = np.ptp(ts[v]/c[v], axis=0)
        rng[rng == 0] = 1.0
        scores.append((rmse / rng * 100).mean())

    return np.mean(scores)

# ============================================================
# 8. CV
# ============================================================
gkf = GroupKFold(5)
scores = []

for f, (tr, va) in enumerate(gkf.split(X, y, groups)):
    Xm, Xs = X[tr].mean((0,1)), X[tr].std((0,1))
    Ym, Ys = y[tr].mean((0,1)), y[tr].std((0,1))
    Xs[Xs==0]=1; Ys[Ys==0]=1

    tr_ds = WindowDataset((X[tr]-Xm)/Xs, (y[tr]-Ym)/Ys,
                          np.array(trial_ids)[tr], np.array(window_starts)[tr])
    va_ds = WindowDataset((X[va]-Xm)/Xs, (y[va]-Ym)/Ys,
                          np.array(trial_ids)[va], np.array(window_starts)[va])

    model = TCN(X.shape[-1], y.shape[-1]).to(device)
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)

    best, wait = 1e9, 0
    for ep in range(num_epochs):
        model.train()
        for xb, yb, _, _ in DataLoader(tr_ds, batch_size, True):
            opt.zero_grad()
            F.mse_loss(model(xb.to(device)), yb.to(device)).backward()
            opt.step()

        sc = trialwise_nrmse_mc(model,
                                DataLoader(va_ds, batch_size, False),
                                Ym, Ys)
        if sc < best:
            best, wait = sc, 0
            torch.save(model.state_dict(), f"{EXP_DIR}/fold{f}.pt")
        else:
            wait += 1
            if wait >= patience:
                break

    scores.append(best)

np.save(f"{EXP_DIR}/cv_scores.npy", np.array(scores))
print("Mean NRMSE (TCN + MC Dropout):", np.mean(scores))
```

```python
import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupKFold
from collections import defaultdict
import pickle

# ---------------------------
# 固定値・環境設定
# ---------------------------
SEED = 30
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

EXP_DIR = "/content/drive/MyDrive/t-teramae/experiment_tcn_mdn"
os.makedirs(EXP_DIR, exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "models"), exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "norm_params"), exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "scores"), exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "history"), exist_ok=True)

# ---------------------------
# データ設定
# ---------------------------
file_path = "/content/new_wf_tr_extraction_selected.csv"
encoding = "cp932"

WINDOW = 1000
STRIDE = 250
batch_size = 8
num_epochs = 1000
lr = 1e-4

hidden_ch = 64
levels = 5
kernel_size = 5
dropout = 0.1
weight_decay = 1e-5
patience = 50

N_MIXTURES = 3

# 入力・出力カラム
input_keys = [
    "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm",
    "左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm",
    "左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm",
    "左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm",
    "左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm",
    "左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm",
    "右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm",
    "右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm",
    "右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm",
    "右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm",
    "右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm",
    "右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm",
    "左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm",
    "左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm",
    "左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm",
    "左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm",
    "左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm",
    "左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm",
    "右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm",
    "右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm",
    "右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm",
    "右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm",
    "PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm",
    "股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm",
    "肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm",
    "体重心Xcm","体重心Ycm","体重心Zcm"
]
output_keys = [
    "左足Fzkg","左足Fxkg","左足Fykg",
    "右足Fzkg","右足Fxkg","右足Fykg"
]
WEIGHT_COL = "体重"

# ---------------------------
# データ読み込み・前処理
# ---------------------------
df = pd.read_csv(file_path, encoding=encoding)
TIME_COL = next(c for c in ["time","Time","timestamp","時間"] if c in df.columns)
df[TIME_COL] = df[TIME_COL].astype(float)

df["trial_id"] = ((df["ID"] != df["ID"].shift()) | (df[TIME_COL].diff() > 0.005)).cumsum()

# WINDOW以上の長さのトライアルのみ抽出
trial_len = df.groupby("trial_id").size()
valid_trials = trial_len[trial_len >= WINDOW].index
df = df[df["trial_id"].isin(valid_trials)].reset_index(drop=True)

# ウィンドウ切り出し
X, y, groups, trial_ids, window_starts = [], [], [], [], []

for tid, g in df.groupby("trial_id"):
    if g[input_keys + output_keys + [WEIGHT_COL]].isna().any().any():
        continue
    Xv = g[input_keys].values.astype(np.float32)
    weight = float(g[WEIGHT_COL].iloc[0])
    yv = g[output_keys].values.astype(np.float32) / weight * 100.0
    sid = g["ID"].iloc[0]
    for s in range(0, len(Xv)-WINDOW+1, STRIDE):
        X.append(Xv[s:s+WINDOW])
        y.append(yv[s:s+WINDOW])
        groups.append(sid)
        trial_ids.append(tid)
        window_starts.append(s)

X = np.stack(X)
y = np.stack(y)
groups = np.array(groups)
trial_ids = np.array(trial_ids)
window_starts = np.array(window_starts)

print("Total windows:", len(X))

# ---------------------------
# Dataset
# ---------------------------
class WindowDataset(Dataset):
    def __init__(self, X, y, tid, wstart):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
        self.tid = tid
        self.wstart = wstart

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i], self.tid[i], self.wstart[i]

# ---------------------------
# モデル定義
# ---------------------------
class TemporalBlock(nn.Module):
    def __init__(self, in_ch, out_ch, k, d, drop):
        super().__init__()
        pad = (k-1)*d
        self.c1 = nn.Conv1d(in_ch, out_ch, k, padding=pad, dilation=d)
        self.c2 = nn.Conv1d(out_ch, out_ch, k, padding=pad, dilation=d)
        self.dp = nn.Dropout(drop)
        self.ds = nn.Conv1d(in_ch, out_ch, 1) if in_ch!=out_ch else None

    def forward(self, x):
        y = self.c1(x)[:,:,:-self.c1.padding[0]]
        y = self.dp(F.relu(y))
        y = self.c2(y)[:,:,:-self.c2.padding[0]]
        y = self.dp(F.relu(y))
        r = x if self.ds is None else self.ds(x)
        return F.relu(y+r)

class MDNLayer(nn.Module):
    def __init__(self, in_ch, out_ch, K, sigma_min=1e-2):
        super().__init__()
        self.K = K
        self.out_ch = out_ch
        self.sigma_min = sigma_min  # 下限を追加
        self.pi = nn.Conv1d(in_ch, K, 1)
        self.mu = nn.Conv1d(in_ch, K*out_ch, 1)
        self.sigma = nn.Conv1d(in_ch, K*out_ch, 1)

    def forward(self, x):
        pi = F.softmax(self.pi(x), dim=1)
        mu = self.mu(x)
        #sigma = torch.exp(self.sigma(x)) + 1e-6
        sigma = F.softplus(self.sigma(x)) + self.sigma_min
        mu = mu.view(x.size(0), self.K, self.out_ch, -1)
        sigma = sigma.view(x.size(0), self.K, self.out_ch, -1)
        return pi, mu, sigma

class TCN_MDN(nn.Module):
    def __init__(self, in_dim, out_dim, K):
        super().__init__()
        layers, ch = [], in_dim
        for i in range(levels):
            layers.append(TemporalBlock(ch, hidden_ch, kernel_size, 2**i, dropout))
            ch = hidden_ch
        self.tcn = nn.Sequential(*layers)
        self.mdn = MDNLayer(hidden_ch, out_dim, K)

    def forward(self, x):
        x = x.transpose(1,2)
        h = self.tcn(x)
        return self.mdn(h)

# ---------------------------
# 損失・推論関数
# ---------------------------
def mdn_loss(pi, mu, sigma, y):
    y = y.permute(0,2,1).unsqueeze(1)
    g = torch.exp(-0.5*((y-mu)/sigma)**2)/(sigma*np.sqrt(2*np.pi))
    p = torch.sum(pi.unsqueeze(2)*g, dim=1)
    return -torch.log(p+1e-8).mean()

def mdn_mean(pi, mu):
    return torch.sum(pi.unsqueeze(2)*mu, dim=1).permute(0,2,1)

def trialwise_nrmse_mdn(model, dataset, ym, ys):
    model.eval()
    loader = DataLoader(dataset, batch_size=8, shuffle=False)
    P, T, S, L = defaultdict(list), defaultdict(list), defaultdict(list), defaultdict(int)
    with torch.no_grad():
        for xb, yb, tid, ws in loader:
            xb = xb.to(device)
            pi, mu, _ = model(xb)
            pred = mdn_mean(pi, mu).cpu().numpy() * ys + ym
            true = yb.numpy() * ys + ym
            for i in range(len(pred)):
                t = int(tid[i])
                s = int(ws[i])
                P[t].append(pred[i])
                T[t].append(true[i])
                S[t].append(s)
                L[t] = max(L[t], s + WINDOW)
    scores = []
    for t in P:
        ps = np.zeros((L[t], y.shape[-1]))
        ts = np.zeros((L[t], y.shape[-1]))
        cnt = np.zeros((L[t], 1))
        for p, tr, s in zip(P[t], T[t], S[t]):
            ps[s:s+WINDOW] += p
            ts[s:s+WINDOW] += tr
            cnt[s:s+WINDOW] += 1
        valid = cnt.squeeze() > 0
        rmse = np.sqrt(np.mean((ps[valid]/cnt[valid] - ts[valid]/cnt[valid])**2, axis=0))
        rng = np.ptp(ts[valid]/cnt[valid], axis=0)
        rng[rng==0] = 1.0
        scores.append((rmse/rng*100).mean())
    return np.mean(scores)

# ---------------------------
# 学習・保存ループ
# ---------------------------
gkf = GroupKFold(5)
fold_scores = []

for f,(tr,va) in enumerate(gkf.split(X,y,groups)):
    print(f"\n=== Fold {f+1} ===")
    Xtr,Xva = X[tr],X[va]
    ytr,yva = y[tr],y[va]

    # 正規化
    Xm,Xs = Xtr.reshape(-1,Xtr.shape[-1]).mean(0), Xtr.reshape(-1,Xtr.shape[-1]).std(0)
    ym,ys = ytr.reshape(-1,ytr.shape[-1]).mean(0), ytr.reshape(-1,ytr.shape[-1]).std(0)
    Xs[Xs==0]=1; ys[ys==0]=1
    Xtr=(Xtr-Xm)/Xs; Xva=(Xva-Xm)/Xs
    ytr=(ytr-ym)/ys; yva=(yva-ym)/ys

    tr_ds = WindowDataset(Xtr,ytr,trial_ids[tr],window_starts[tr])
    va_ds = WindowDataset(Xva,yva,trial_ids[va],window_starts[va])

    model = TCN_MDN(X.shape[-1], y.shape[-1], N_MIXTURES).to(device)
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)

    best, wait = np.inf, 0
    history = {"loss": [], "val_nrmse": []}

    for ep in range(num_epochs):
        model.train()
        epoch_loss = 0
        for xb, yb, _, _ in DataLoader(tr_ds, batch_size, True):
            xb, yb = xb.to(device), yb.to(device)
            pi, mu, sigma = model(xb)
            loss = mdn_loss(pi, mu, sigma, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            epoch_loss += loss.item() * xb.size(0)
        epoch_loss /= len(tr_ds)
        val_score = trialwise_nrmse_mdn(model, va_ds, ym, ys)
        history["loss"].append(epoch_loss)
        history["val_nrmse"].append(val_score)

        if val_score < best:
            best, wait = val_score, 0
            torch.save(model.state_dict(), os.path.join(EXP_DIR, f"models/model_fold{f+1}.pt"))
            with open(os.path.join(EXP_DIR, f"norm_params/norm_fold{f+1}.pkl"), "wb") as f_norm:
                pickle.dump({"Xm": Xm, "Xs": Xs, "ym": ym, "ys": ys}, f_norm)
            print(f"Epoch {ep+1:03d} | loss {epoch_loss:.4f} | val NRMSE {val_score:.3f}")
        else:
            wait += 1
            if wait >= patience:
                break


    fold_scores.append(best)
    with open(os.path.join(EXP_DIR, f"scores/score_fold{f+1}.pkl"), "wb") as f_score:
        pickle.dump(best, f_score)
    with open(os.path.join(EXP_DIR, f"history/history_fold{f+1}.pkl"), "wb") as f_hist:
        pickle.dump(history, f_hist)

mean_score = np.mean(fold_scores)
print("\nMean NRMSE:", mean_score)
with open(os.path.join(EXP_DIR, "scores/fold_scores.pkl"), "wb") as f_allscore:
    pickle.dump(fold_scores, f_allscore)
```

```python
import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupKFold
from collections import defaultdict
import pickle
import gpytorch

# ---------------------------
# 固定値・環境設定
# ---------------------------
SEED = 30
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

EXP_DIR = "/content/drive/MyDrive/t-teramae/experiment_tcn_spgp"
os.makedirs(EXP_DIR, exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "models"), exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "norm_params"), exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "scores"), exist_ok=True)
os.makedirs(os.path.join(EXP_DIR, "history"), exist_ok=True)

# ---------------------------
# データ設定
# ---------------------------
file_path = "/content/new_wf_tr_extraction_selected.csv"
encoding = "cp932"

WINDOW = 1000
STRIDE = 250
batch_size = 8
num_epochs = 1000
lr = 1e-4

hidden_ch = 64
levels = 5
kernel_size = 5
dropout = 0.1
weight_decay = 1e-5
patience = 50

# 入力・出力カラム
input_keys = [
    "左肩峰Xcm","左肩峰Ycm","左肩峰Zcm","左腸骨陵Xcm","左腸骨陵Ycm","左腸骨陵Zcm",
    "左大転子Xcm","左大転子Ycm","左大転子Zcm","左膝関節外Xcm","左膝関節外Ycm",
    "左膝関節外Zcm","左膝関節内Xcm","左膝関節内Ycm","左膝関節内Zcm",
    "左足関節外Xcm","左足関節外Ycm","左足関節外Zcm","左足関節内Xcm",
    "左足関節内Ycm","左足関節内Zcm","左中足骨外Xcm","左中足骨外Ycm",
    "左中足骨外Zcm","左中足骨内Xcm","左中足骨内Ycm","左中足骨内Zcm",
    "右肩峰Xcm","右肩峰Ycm","右肩峰Zcm","右腸骨陵Xcm","右腸骨陵Ycm","右腸骨陵Zcm",
    "右大転子Xcm","右大転子Ycm","右大転子Zcm","右膝関節外Xcm","右膝関節外Ycm",
    "右膝関節外Zcm","右膝関節内Xcm","右膝関節内Ycm","右膝関節内Zcm",
    "右足関節外Xcm","右足関節外Ycm","右足関節外Zcm","右足関節内Xcm",
    "右足関節内Ycm","右足関節内Zcm","右中足骨外Xcm","右中足骨外Ycm",
    "右中足骨外Zcm","右中足骨内Xcm","右中足骨内Ycm","右中足骨内Zcm",
    "左PSISXcm","左PSISYcm","左PSISZcm","右PSISXcm","右PSISYcm","右PSISZcm",
    "左踵Xcm","左踵Ycm","左踵Zcm","右踵Xcm","右踵Ycm","右踵Zcm",
    "左股関節点推定Xcm","左股関節点推定Ycm","左股関節点推定Zcm",
    "左膝関節点推定Xcm","左膝関節点推定Ycm","左膝関節点推定Zcm",
    "左足関節点推定Xcm","左足関節点推定Ycm","左足関節点推定Zcm",
    "左中足骨点推定Xcm","左中足骨点推定Ycm","左中足骨点推定Zcm",
    "右股関節点推定Xcm","右股関節点推定Ycm","右股関節点推定Zcm",
    "右膝関節点推定Xcm","右膝関節点推定Ycm","右膝関節点推定Zcm",
    "右足関節点推定Xcm","右足関節点推定Ycm","右足関節点推定Zcm",
    "右中足骨点推定Xcm","右中足骨点推定Ycm","右中足骨点推定Zcm",
    "PSIS中点Xcm","PSIS中点Ycm","PSIS中点Zcm",
    "股関節投影点中点Xcm","股関節投影点中点Ycm","股関節投影点中点Zcm",
    "肩関節中点Xcm","肩関節中点Ycm","肩関節中点Zcm",
    "体重心Xcm","体重心Ycm","体重心Zcm"
]
output_keys = [
    "左足Fzkg","左足Fxkg","左足Fykg",
    "右足Fzkg","右足Fxkg","右足Fykg"
]
WEIGHT_COL = "体重"

# ---------------------------
# データ読み込み・前処理
# ---------------------------
df = pd.read_csv(file_path, encoding=encoding)
TIME_COL = next(c for c in ["time","Time","timestamp","時間"] if c in df.columns)
df[TIME_COL] = df[TIME_COL].astype(float)
df["trial_id"] = ((df["ID"] != df["ID"].shift()) | (df[TIME_COL].diff() > 0.005)).cumsum()

trial_len = df.groupby("trial_id").size()
valid_trials = trial_len[trial_len >= WINDOW].index
df = df[df["trial_id"].isin(valid_trials)].reset_index(drop=True)

X, y, groups, trial_ids, window_starts = [], [], [], [], []

for tid, g in df.groupby("trial_id"):
    if g[input_keys + output_keys + [WEIGHT_COL]].isna().any().any():
        continue
    Xv = g[input_keys].values.astype(np.float32)
    weight = float(g[WEIGHT_COL].iloc[0])
    yv = g[output_keys].values.astype(np.float32) / weight * 100.0
    sid = g["ID"].iloc[0]
    for s in range(0, len(Xv)-WINDOW+1, STRIDE):
        X.append(Xv[s:s+WINDOW])
        y.append(yv[s:s+WINDOW])
        groups.append(sid)
        trial_ids.append(tid)
        window_starts.append(s)

X = np.stack(X)
y = np.stack(y)
groups = np.array(groups)
trial_ids = np.array(trial_ids)
window_starts = np.array(window_starts)
print("Total windows:", len(X))

# ---------------------------
# Dataset
# ---------------------------
class WindowDataset(Dataset):
    def __init__(self, X, y, tid, wstart):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
        self.tid = tid
        self.wstart = wstart

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i], self.tid[i], self.wstart[i]

# ---------------------------
# NRMSE関数
# ---------------------------
def nrmse(y_true, y_pred):
    rmse = np.sqrt(np.mean((y_true - y_pred)**2, axis=0))
    rng = np.ptp(y_true, axis=0)
    # rng がスカラーか配列かで処理を分ける
    if np.isscalar(rng):
        if rng == 0:
            rng = 1.0
    else:
        rng[rng == 0] = 1.0
    return (rmse / rng).mean() * 100

# ---------------------------
# TCNモデル
# ---------------------------
class TemporalBlock(nn.Module):
    def __init__(self, in_ch, out_ch, k, d, drop):
        super().__init__()
        pad = (k-1)*d
        self.c1 = nn.Conv1d(in_ch, out_ch, k, padding=pad, dilation=d)
        self.c2 = nn.Conv1d(out_ch, out_ch, k, padding=pad, dilation=d)
        self.dp = nn.Dropout(drop)
        self.ds = nn.Conv1d(in_ch, out_ch, 1) if in_ch!=out_ch else None

    def forward(self, x):
        y = self.c1(x)[:,:,:-self.c1.padding[0]]
        y = self.dp(F.relu(y))
        y = self.c2(y)[:,:,:-self.c2.padding[0]]
        y = self.dp(F.relu(y))
        r = x if self.ds is None else self.ds(x)
        return F.relu(y+r)

class TCN(nn.Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        layers, ch = [], in_dim
        for i in range(levels):
            layers.append(TemporalBlock(ch, hidden_ch, kernel_size, 2**i, dropout))
            ch = hidden_ch
        self.tcn = nn.Sequential(*layers)
        self.fc = nn.Linear(hidden_ch, out_dim)

    def forward(self, x):
        x = x.transpose(1,2)
        h = self.tcn(x)
        h_mean = h.mean(dim=2)
        return self.fc(h_mean)

# ---------------------------
# SPGPモデル
# ---------------------------
class SparseGPModel(gpytorch.models.ApproximateGP):
    def __init__(self, inducing_points):
        variational_distribution = gpytorch.variational.CholeskyVariationalDistribution(
            inducing_points.size(0))
        variational_strategy = gpytorch.variational.VariationalStrategy(
            self, inducing_points, variational_distribution, learn_inducing_locations=True)
        super().__init__(variational_strategy)
        self.mean_module = gpytorch.means.ZeroMean()
        self.covar_module = gpytorch.kernels.ScaleKernel(
            gpytorch.kernels.RBFKernel()
        )

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)

# ---------------------------
# Fold学習
# ---------------------------
gkf = GroupKFold(5)

for f, (tr, va) in enumerate(gkf.split(X, y, groups)):
    print(f"\n=== Fold {f+1} ===")
    Xtr, Xva = X[tr], X[va]
    ytr, yva = y[tr], y[va]

    # --- 正規化 ---
    Xm, Xs = Xtr.reshape(-1,Xtr.shape[-1]).mean(0), Xtr.reshape(-1,Xtr.shape[-1]).std(0)
    Xs[Xs==0] = 1
    Xtr = (Xtr - Xm)/Xs
    Xva = (Xva - Xm)/Xs

    # --- TCNモデル作成 ---
    tcn = TCN(X.shape[-1], y.shape[-1]).to(device)
    tcn_opt = torch.optim.AdamW(tcn.parameters(), lr=lr, weight_decay=weight_decay)
    mse_loss = nn.MSELoss()

    # ウィンドウ平均
    ytr_avg = ytr.mean(axis=1)
    yva_avg = yva.mean(axis=1)

    tr_ds = WindowDataset(Xtr, ytr_avg, trial_ids[tr], window_starts[tr])
    va_ds = WindowDataset(Xva, yva_avg, trial_ids[va], window_starts[va])

    # --- TCN学習 ---
    best_val_nrmse = np.inf
    wait = 0

    for ep in range(num_epochs):
        tcn.train()
        epoch_loss = 0
        for xb, yb, _, _ in DataLoader(tr_ds, batch_size, True):
            xb, yb = xb.to(device), yb.to(device)
            y_pred = tcn(xb)
            loss = mse_loss(y_pred, yb)
            tcn_opt.zero_grad()
            loss.backward()
            tcn_opt.step()
            epoch_loss += loss.item() * xb.size(0)
        epoch_loss /= len(tr_ds)

        # Validation
        tcn.eval()
        va_preds = []
        va_trues = []
        with torch.no_grad():
            for xb, yb, _, _ in DataLoader(va_ds, batch_size, False):
                xb = xb.to(device)
                y_pred = tcn(xb).cpu().numpy()
                va_preds.append(y_pred)
                va_trues.append(yb.numpy())
        va_preds = np.vstack(va_preds)
        va_trues = np.vstack(va_trues)
        val_nrmse = nrmse(va_trues, va_preds)
        print(f"Epoch {ep+1} | train_loss {epoch_loss:.4f} | val NRMSE {val_nrmse:.3f}")

        # ベストモデル更新
        if val_nrmse < best_val_nrmse:
            best_val_nrmse = val_nrmse
            wait = 0
            torch.save(tcn.state_dict(), os.path.join(EXP_DIR, f"models/tcn_fold{f+1}_best.pt"))
        else:
            wait += 1
            if wait >= patience:
                break

    # --- SPGP学習用にベストTCNをロード ---
    tcn.load_state_dict(torch.load(os.path.join(EXP_DIR, f"models/tcn_fold{f+1}_best.pt")))
    tcn.eval()

    with torch.no_grad():
        # ベストTCNの出力をSPGP入力に
        Xtr_tcn = tcn(torch.tensor(Xtr, dtype=torch.float32).to(device)).cpu()
        ytr_residual = torch.tensor(ytr_avg, dtype=torch.float32) - Xtr_tcn

    # SPGP 残差学習（バリデーションNRMSEでアーリーストップ）
    num_inducing = min(500, Xtr_tcn.size(0))
    inducing_idx = np.random.choice(Xtr_tcn.size(0), num_inducing, replace=False)
    inducing_points = Xtr_tcn[inducing_idx, :]

    # バリデーション TCN 出力
    with torch.no_grad():
        Xva_tcn = tcn(torch.tensor(Xva, dtype=torch.float32).to(device)).cpu()
        yva_avg = yva.mean(axis=1)

    gp_models = []
    likelihoods = []
    gp_patience = 50
    gp_max_steps = 1000
    num_outputs = ytr_residual.shape[1]

    for i in range(num_outputs):
        likelihood = gpytorch.likelihoods.GaussianLikelihood().to(device)
        gp_model = SparseGPModel(inducing_points.to(device)).to(device)
        gp_model.train(); likelihood.train()

        gp_opt = torch.optim.Adam(list(gp_model.parameters()) + list(likelihood.parameters()), lr=0.01)
        train_y_i = ytr_residual[:, i].to(device)
        mll = gpytorch.mlls.VariationalELBO(likelihood, gp_model, num_data=train_y_i.size(0))

        best_nrmse = np.inf
        wait = 0

        for step in range(gp_max_steps):
            gp_model.train(); likelihood.train()
            gp_opt.zero_grad()
            output = gp_model(Xtr_tcn.to(device))
            loss = -mll(output, train_y_i)
            loss.backward()
            gp_opt.step()

            # --- バリデーションNRMSEで早期終了 ---
            gp_model.eval()
            likelihood.eval()
            with torch.no_grad(), gpytorch.settings.fast_pred_var():
                # 既に Xva_tcn は CPU 上の tensor なら、まず GPU に移動
                Xva_tcn_device = Xva_tcn.to(device)
                gp_val_pred = likelihood(gp_model(Xva_tcn_device)).mean.cpu().numpy()
            current_nrmse = nrmse(yva_avg[:, i], gp_val_pred)

            if step % 10 == 0:
                print(f"GP output {i} | step {step} | ELBO loss {loss.item():.4f} | val NRMSE {current_nrmse:.3f}")

            if current_nrmse < best_nrmse:
                best_nrmse = current_nrmse
                wait = 0
                best_state_dict = {
                    'model': gp_model.state_dict(),
                    'likelihood': likelihood.state_dict()
                }
            else:
                wait += 1
                if wait >= gp_patience:
                    print(f"GP output {i} early stopping at step {step}")
                    gp_model.load_state_dict(best_state_dict['model'])
                    likelihood.load_state_dict(best_state_dict['likelihood'])
                    break

        gp_models.append(gp_model)
        likelihoods.append(likelihood)


    # --- Validation: TCN+SPGP ---
    tcn.eval()
    va_preds_final = []
    with torch.no_grad(), gpytorch.settings.fast_pred_var():
        for xb, yb, _, _ in DataLoader(va_ds, batch_size, False):
            xb_device = xb.to(device)
            tcn_pred = tcn(xb_device)  # [B, 6]
            gp_pred_total = torch.zeros_like(tcn_pred)

            for i in range(num_outputs):
                gp_model = gp_models[i].eval()
                likelihood = likelihoods[i].eval()
                gp_pred_total[:, i] = likelihood(gp_model(tcn_pred)).mean

            va_preds_final.append((tcn_pred + gp_pred_total).cpu().numpy())

    va_preds_final = np.vstack(va_preds_final)
    final_nrmse = nrmse(va_trues, va_preds_final)
    print(f"Fold {f+1} final NRMSE (TCN+SPGP): {final_nrmse:.3f}")

print("All done!")
```

```python
# ============================================================
# MDN-TCN Visualization (Single Cell Complete Version)
# ============================================================

import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torch.utils.data import Dataset

# ----------------------------
# device
# ----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ----------------------------
# paths & hyperparameters
# ----------------------------
BASE_DIR = "/content/drive/MyDrive/t-teramae/experiment_tcn_mdn"

WINDOW = 1000
hidden_ch = 64
levels = 5
kernel_size = 5
dropout = 0.1
N_MIXTURES = 3

# ============================================================
# Dataset
# ============================================================
class WindowDataset(Dataset):
    def __init__(self, X, y, tid, wstart):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
        self.tid = tid
        self.wstart = wstart

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i], self.tid[i], self.wstart[i]

# ============================================================
# Model definition (TCN + MDN)
# ============================================================
class TemporalBlock(nn.Module):
    def __init__(self, in_ch, out_ch, k, d, drop):
        super().__init__()
        pad = (k - 1) * d
        self.c1 = nn.Conv1d(in_ch, out_ch, k, padding=pad, dilation=d)
        self.c2 = nn.Conv1d(out_ch, out_ch, k, padding=pad, dilation=d)
        self.dp = nn.Dropout(drop)
        self.ds = nn.Conv1d(in_ch, out_ch, 1) if in_ch != out_ch else None

    def forward(self, x):
        y = self.c1(x)[:, :, :-self.c1.padding[0]]
        y = self.dp(F.relu(y))
        y = self.c2(y)[:, :, :-self.c2.padding[0]]
        y = self.dp(F.relu(y))
        r = x if self.ds is None else self.ds(x)
        return F.relu(y + r)

class MDNLayer(nn.Module):
    def __init__(self, in_ch, out_ch, K):
        super().__init__()
        self.K = K
        self.out_ch = out_ch
        self.pi = nn.Conv1d(in_ch, K, 1)
        self.mu = nn.Conv1d(in_ch, K * out_ch, 1)
        self.sigma = nn.Conv1d(in_ch, K * out_ch, 1)

    def forward(self, x):
        pi = F.softmax(self.pi(x), dim=1)
        mu = self.mu(x)
        sigma = torch.exp(self.sigma(x)) + 1e-6
        mu = mu.view(x.size(0), self.K, self.out_ch, -1)
        sigma = sigma.view(x.size(0), self.K, self.out_ch, -1)
        return pi, mu, sigma

class TCN_MDN(nn.Module):
    def __init__(self, in_dim, out_dim, K):
        super().__init__()
        layers, ch = [], in_dim
        for i in range(levels):
            layers.append(
                TemporalBlock(ch, hidden_ch, kernel_size, 2 ** i, dropout)
            )
            ch = hidden_ch
        self.tcn = nn.Sequential(*layers)
        self.mdn = MDNLayer(hidden_ch, out_dim, K)

    def forward(self, x):
        x = x.transpose(1, 2)
        h = self.tcn(x)
        return self.mdn(h)

# ============================================================
# Load saved data
# ============================================================
X = np.load(os.path.join(BASE_DIR, "X.npy"))
y = np.load(os.path.join(BASE_DIR, "y.npy"))
trial_ids = np.load(os.path.join(BASE_DIR, "trial_ids.npy"))
window_starts = np.load(os.path.join(BASE_DIR, "window_starts.npy"))

norm = np.load(os.path.join(BASE_DIR, "norm_params_fold0.npz"))
ym, ys = norm["y_mean"], norm["y_std"]

dataset = WindowDataset(X, y, trial_ids, window_starts)

# ============================================================
# Load model
# ============================================================
model = TCN_MDN(
    in_dim=X.shape[-1],
    out_dim=y.shape[-1],
    K=N_MIXTURES
).to(device)

model.load_state_dict(
    torch.load(os.path.join(BASE_DIR, "model_mdn_fold0.pt"),
               map_location=device)
)
model.eval()

# ============================================================
# Visualization function
# ============================================================
def visualize_mdn_trial(
    model, dataset, trial_id,
    ym, ys,
    out_ch=0,
    K=3,
    title="MDN-TCN Output"
):
    model.eval()
    idxs = [i for i in range(len(dataset)) if dataset.tid[i] == trial_id]
    assert len(idxs) > 0, "trial_id not found"

    preds_mu = [[] for _ in range(K)]
    preds_pi = [[] for _ in range(K)]
    preds_mean, trues, starts = [], [], []

    with torch.no_grad():
        for i in idxs:
            xb, yb, _, ws = dataset[i]
            xb = xb.unsqueeze(0).to(device)

            pi, mu, _ = model(xb)

            mu = mu.cpu().numpy()[0] * ys + ym
            pi = pi.cpu().numpy()[0]
            yb = yb.numpy() * ys + ym

            for k in range(K):
                preds_mu[k].append(mu[k, out_ch])
                preds_pi[k].append(pi[k])

            preds_mean.append(
                np.sum(pi[:, None, :] * mu[:, out_ch:out_ch + 1, :], axis=0).squeeze()
            )
            trues.append(yb[:, out_ch])
            starts.append(ws)

    T = max(starts) + WINDOW
    gt = np.zeros(T)
    mean_pred = np.zeros(T)
    mu_k = [np.zeros(T) for _ in range(K)]
    pi_k = [np.zeros(T) for _ in range(K)]
    cnt = np.zeros(T)

    for i in range(len(starts)):
        s = starts[i]
        gt[s:s + WINDOW] += trues[i]
        mean_pred[s:s + WINDOW] += preds_mean[i]
        for k in range(K):
            mu_k[k][s:s + WINDOW] += preds_mu[k][i]
            pi_k[k][s:s + WINDOW] += preds_pi[k][i]
        cnt[s:s + WINDOW] += 1

    valid = cnt > 0
    gt[valid] /= cnt[valid]
    mean_pred[valid] /= cnt[valid]
    for k in range(K):
        mu_k[k][valid] /= cnt[valid]
        pi_k[k][valid] /= cnt[valid]

    fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

    axes[0].plot(gt, "k", lw=2, label="Ground Truth")
    axes[0].plot(mean_pred, "b", lw=2, label="MDN Mean")

    colors = ["r", "g", "orange"]
    for k in range(K):
        axes[0].plot(mu_k[k], "--", color=colors[k], label=f"$\\mu_{k+1}$")

    axes[0].set_ylabel("GRF (%BW)")
    axes[0].legend()

    for k in range(K):
        axes[1].plot(pi_k[k], color=colors[k], label=f"$\\pi_{k+1}$")

    axes[1].set_ylabel("Mixture Weight")
    axes[1].set_xlabel("Time (samples)")
    axes[1].set_ylim(0, 1)
    axes[1].legend()

    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# ============================================================
# Run visualization
# ============================================================
trial_id_example = dataset.tid[0]

visualize_mdn_trial(
    model=model,
    dataset=dataset,
    trial_id=trial_id_example,
    ym=ym,
    ys=ys,
    out_ch=0,
    K=N_MIXTURES,
    title="Representative Post-Stroke Trial (Left Fz)"
)
```
